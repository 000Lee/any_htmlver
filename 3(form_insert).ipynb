```python
import json

# 첫 번째 JSON 파일 로드 (리스트 형태)
#컨트롤러
with open('three_docbody_insert.json', 'r', encoding='utf-8') as f:
    first_data = json.load(f)

# 두 번째 JSON 파일 로드 (딕셔너리 형태)
#컨트롤러
with open('one_formCrawling_result.json', 'r', encoding='utf-8') as f:
    second_data = json.load(f)

# formName 매칭 작업
matched_count = 0
for document in first_data:
    source_id = document.get('sourceId')
    
    # sourceId가 두 번째 데이터의 키에 존재하는지 확인
    if source_id and source_id in second_data:
        # 양식명 값을 formName에 입력
        document['formName'] = second_data[source_id]['양식명']
        matched_count += 1
        print(f"매칭 완료: {source_id} -> {document['formName']}")

print(f"\n총 {matched_count}개 문서의 formName이 업데이트되었습니다.")

# 결과를 새 파일로 저장
#컨트롤러
with open('three_form_insert.json', 'w', encoding='utf-8') as f:
    json.dump(first_data, f, ensure_ascii=False, indent=2)

print("결과 파일이 저장되었습니다: three_form_insert.json")
```

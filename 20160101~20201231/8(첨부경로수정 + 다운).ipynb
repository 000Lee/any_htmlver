```python
# ì²¨ë¶€íŒŒì¼ ê²½ë¡œ ë³€í™˜ - any_approval DB ì „ìê²°ì¬ ë¬¸ì„œìš©
# documents í…Œì´ë¸”ì˜ attaches ì»¬ëŸ¼ ì²˜ë¦¬
# end_yearë¡œ ì—°ë„ë³„ í•„í„°ë§

import json
import os
import shutil
from pathlib import Path
import pymysql
from sqlalchemy import create_engine

# ==================== DB ì„¤ì • ====================
USER = "root"
PASS = "1234"
HOST = "127.0.0.1"
PORT = 3306
DB = "any_approval"  # âœ¨ ë³€ê²½

# ==================== ì„¤ì • ì˜ì—­ (ì—¬ê¸°ë¥¼ ìˆ˜ì •í•˜ì„¸ìš”) ====================

# ì¢…ë£Œì—°ë„ (í•„í„°ë§ìš©)
END_YEAR = 2020  # âœ¨ ë³€ê²½: 2025, 2024, 2023 ë“±

# ì²¨ë¶€íŒŒì¼ ì›ë³¸ ê²½ë¡œë“¤ - ì—¬ëŸ¬ ê²½ë¡œ ì¶”ê°€ ê°€ëŠ¥
BASE_FILE_PATHS = [
    # r"C:\Users\LEEJUHWAN\Downloads\2021-01-01~2025-10-31\html\resource\file\apr",  # âœ¨ brd â†’ apr
    r"C:\Users\LEEJUHWAN\Downloads\2016-01-01~2020-12-31\html\resource\file\apr",  # âœ¨ brd â†’ apr
    # r"C:\Users\LEEJUHWAN\Downloads\2011-01-01~2015-12-31\html\resource\file\apr",  # âœ¨ brd â†’ apr
    # r"C:\Users\LEEJUHWAN\Downloads\2010-01-01~2010-12-31\html\resource\file\apr",  # âœ¨ brd â†’ apr
    # ë” ì¶”ê°€í•˜ê³  ì‹¶ìœ¼ë©´ ì—¬ê¸°ì— ê³„ì† ì¶”ê°€
]

# ì²¨ë¶€íŒŒì¼ ë³µì‚¬ë³¸ ì €ì¥ í´ë” (ì—°ë„ë³„ë¡œ ìë™ ìƒì„±)
OUTPUT_FILE_FOLDER = rf"C:\Users\LEEJUHWAN\Downloads\approval_{END_YEAR}_attachments"  # âœ¨ ë³€ê²½

# pathì—ì„œ ì œê±°í•  ì ‘ë‘ì‚¬ (ì´ ë¶€ë¶„ì„ ì œê±°í•˜ê³  ìƒëŒ€ê²½ë¡œë¡œ ë³€ê²½)
REMOVE_PREFIX = r"C:\Users\LEEJUHWAN\Downloads"

# ======================================================================


def find_attachment_folder(doc_id):
    """ì—¬ëŸ¬ ê²½ë¡œì—ì„œ ì²¨ë¶€íŒŒì¼ í´ë” ì°¾ê¸°"""
    for base_path in BASE_FILE_PATHS:
        attachment_folder = Path(base_path) / doc_id
        if attachment_folder.exists():
            return attachment_folder
    return None


def find_file_in_folder(folder, filename):
    """í´ë” ë‚´ì—ì„œ íŒŒì¼ëª…ê³¼ ì¼ì¹˜í•˜ëŠ” íŒŒì¼ ì°¾ê¸° (ëŒ€ì†Œë¬¸ì êµ¬ë¶„ ì—†ìŒ)"""
    if not folder or not folder.exists():
        return None
    
    # ì •í™•í•œ íŒŒì¼ëª…ìœ¼ë¡œ ë¨¼ì € ì°¾ê¸°
    target_file = folder / filename
    if target_file.exists() and target_file.is_file():
        return target_file
    
    # ëŒ€ì†Œë¬¸ì êµ¬ë¶„ ì—†ì´ ì°¾ê¸°
    for file in folder.iterdir():
        if file.is_file() and file.name.lower() == filename.lower():
            return file
    
    return None


def process_single_document_attachments(data, doc_index):
    """í•˜ë‚˜ì˜ documentì˜ ì²¨ë¶€íŒŒì¼ ì²˜ë¦¬"""
    
    # source_idì—ì„œ ë¬¸ì„œID ì¶”ì¶œ
    source_id = data.get('source_id', '')  # âœ¨ sourceId â†’ source_id
    if not source_id:
        print(f"âš ï¸ [{doc_index}] source_id ì—†ìŒ, ê±´ë„ˆëœ€")
        return data
    
    doc_id = source_id  # âœ¨ lstrip('_') ì œê±°, ê·¸ëŒ€ë¡œ ì‚¬ìš©
    
    # attaches ë°°ì—´ í™•ì¸
    attaches = data.get('attaches', [])
    if not attaches:
        return data  # ì²¨ë¶€íŒŒì¼ ì—†ìœ¼ë©´ ê·¸ëŒ€ë¡œ ë°˜í™˜
    
    # ì—¬ëŸ¬ ê²½ë¡œì—ì„œ ì²¨ë¶€íŒŒì¼ í´ë” ì°¾ê¸°
    attachment_folder = find_attachment_folder(doc_id)
    
    if not attachment_folder:
        print(f"âš ï¸ [{doc_index}] ë¬¸ì„œID {doc_id} í´ë” ì—†ìŒ (ëª¨ë“  ê²½ë¡œ ê²€ìƒ‰ ì™„ë£Œ)")
        # í´ë”ê°€ ì—†ìœ¼ë©´ ëª¨ë“  ì²¨ë¶€íŒŒì¼ì˜ pathë¥¼ ê³µë€ìœ¼ë¡œ
        for attach in attaches:
            attach['path'] = ''
        return data
    
    # ë³µì‚¬ë³¸ ì €ì¥í•  ë¬¸ì„œë³„ í´ë” ìƒì„±
    output_doc_folder = Path(OUTPUT_FILE_FOLDER) / f"apr{doc_id}"  # âœ¨ brd â†’ apr
    output_doc_folder.mkdir(parents=True, exist_ok=True)
    
    print(f"\nğŸ“„ [{doc_index}] ë¬¸ì„œID: {doc_id}")
    print(f"   ì›ë³¸: {attachment_folder}")
    print(f"   ë³µì‚¬: {output_doc_folder}")
    print(f"   âœ… ì²¨ë¶€íŒŒì¼ {len(attaches)}ê°œ ë°œê²¬")
    
    # ê° ì²¨ë¶€íŒŒì¼ ì²˜ë¦¬
    processed_attaches = []
    copied_count = 0
    
    for idx, attach in enumerate(attaches):
        filename = attach.get('name', '')
        if not filename:
            print(f"      âš ï¸ [{idx}] íŒŒì¼ëª… ì—†ìŒ - path ê³µë€ ì²˜ë¦¬")
            attach['path'] = ''
            processed_attaches.append(attach)
            continue
        
        # ì›ë³¸ í´ë”ì—ì„œ íŒŒì¼ ì°¾ê¸°
        source_file = find_file_in_folder(attachment_folder, filename)
        
        if source_file:
            # ë³µì‚¬ë³¸ íŒŒì¼ ê²½ë¡œ
            dest_file = output_doc_folder / filename
            
            # íŒŒì¼ ë³µì‚¬
            try:
                shutil.copy2(source_file, dest_file)
                copied_count += 1
                print(f"      âœ… {filename}")
                
                # pathë¥¼ ìƒëŒ€ê²½ë¡œë¡œ ë³€ê²½ (REMOVE_PREFIX ì œê±°)
                new_path = str(dest_file).replace('\\', '/')
                if REMOVE_PREFIX:
                    new_path = new_path.replace(REMOVE_PREFIX.replace('\\', '/'), '').lstrip('/')
                    new_path = new_path.replace(REMOVE_PREFIX, '').lstrip('\\')
                
                attach['path'] = new_path
                
            except Exception as e:
                print(f"      âŒ {filename} ë³µì‚¬ ì‹¤íŒ¨: {e}")
                attach['path'] = ''
        else:
            print(f"      âš ï¸ íŒŒì¼ ì—†ìŒ: {filename} - path ê³µë€ ì²˜ë¦¬")
            attach['path'] = ''
        
        processed_attaches.append(attach)
    
    print(f"   ğŸ“Š ë³µì‚¬ ì™„ë£Œ: {copied_count}/{len(attaches)}ê°œ")
    
    # ìˆ˜ì •ëœ attaches ì ìš©
    data['attaches'] = processed_attaches
    return data


def process_db(end_year):
    """DBì—ì„œ íŠ¹ì • end_yearì˜ ë°ì´í„°ë§Œ ì½ê³  ì²˜ë¦¬ í›„ DBì— ì—…ë°ì´íŠ¸"""
    
    # ì¶œë ¥ í´ë” ìƒì„±
    os.makedirs(OUTPUT_FILE_FOLDER, exist_ok=True)
    
    # DB ì—°ê²°
    print(f"ğŸ“‚ DB ì—°ê²° ì¤‘...")
    print(f"ğŸ“‚ ì¢…ë£Œì—°ë„: {end_year}")  # âœ¨ ë³€ê²½
    print(f"ğŸ“‚ ê²€ìƒ‰í•  ì²¨ë¶€íŒŒì¼ ê²½ë¡œ: {len(BASE_FILE_PATHS)}ê°œ\n")
    
    connection = pymysql.connect(
        host=HOST,
        user=USER,
        password=PASS,
        database=DB,
        port=PORT,
        charset='utf8mb4'
    )
    
    try:
        with connection.cursor(pymysql.cursors.DictCursor) as cursor:
            # 1. DBì—ì„œ íŠ¹ì • end_yearì˜ ë°ì´í„°ë§Œ ì½ê¸°
            query = "SELECT * FROM documents WHERE end_year = %s"  # âœ¨ ë³€ê²½
            cursor.execute(query, (end_year,))
            documents = cursor.fetchall()  # âœ¨ posts â†’ documents
            
            total_count = len(documents)
            print(f"âœ… ì´ {total_count}ê°œì˜ documents ë°œê²¬\n")
            
            # 2. ê° document ì²˜ë¦¬
            success_count = 0
            total_attachments = 0
            
            for idx, doc in enumerate(documents, 1):  # âœ¨ post â†’ doc
                try:
                    # âœ¨ attaches íŒŒì‹± (ì´ë¯¸ JSONì´ë©´ ê·¸ëŒ€ë¡œ, ë¬¸ìì—´ì´ë©´ loads)
                    attaches_data = doc.get('attaches', [])
                    if isinstance(attaches_data, str):
                        doc['attaches'] = json.loads(attaches_data)
                    else:
                        doc['attaches'] = attaches_data if attaches_data else []
                    
                    # ì²¨ë¶€íŒŒì¼ì´ ìˆëŠ” ê²½ìš°ë§Œ ì¹´ìš´íŠ¸
                    if doc.get('attaches'):
                        total_attachments += len(doc.get('attaches'))
                    
                    # ê¸°ì¡´ ë¡œì§ ê·¸ëŒ€ë¡œ ì‚¬ìš©
                    result = process_single_document_attachments(doc, idx)
                    
                    # attachesê°€ ë³€ê²½ëœ ê²½ìš° DB ì—…ë°ì´íŠ¸
                    if result.get('attaches'):
                        # attachesë¥¼ JSON ë¬¸ìì—´ë¡œ ë³€í™˜
                        attaches_json = json.dumps(result['attaches'], ensure_ascii=False)
                        
                        update_query = """
                            UPDATE documents 
                            SET attaches = %s 
                            WHERE source_id = %s AND end_year = %s
                        """  # âœ¨ board_posts â†’ documents, attaches_json â†’ attaches, sourceId â†’ source_id, type â†’ end_year
                        cursor.execute(update_query, (attaches_json, result['source_id'], end_year))
                        success_count += 1
                    
                except Exception as e:
                    print(f"âŒ [Document {idx}] ì²˜ë¦¬ ì˜¤ë¥˜: {e}")
                    continue
            
            # ë³€ê²½ì‚¬í•­ ì»¤ë°‹
            connection.commit()
            
            print(f"\n{'='*60}")
            print(f"âœ… ì²˜ë¦¬ ì™„ë£Œ!")
            print(f"   - ì¢…ë£Œì—°ë„: {end_year}")  # âœ¨ ë³€ê²½
            print(f"   - ì²¨ë¶€íŒŒì¼ ìˆëŠ” documents: {success_count}ê°œ")
            print(f"   - ì´ ì²¨ë¶€íŒŒì¼ ê°œìˆ˜: {total_attachments}ê°œ")
            print(f"   - ë³µì‚¬ëœ íŒŒì¼ ìœ„ì¹˜: {OUTPUT_FILE_FOLDER}")
            print(f"   - DB í…Œì´ë¸”: documents (attaches ì»¬ëŸ¼ ì—…ë°ì´íŠ¸)")  # âœ¨ ë³€ê²½
            print(f"{'='*60}")
    
    finally:
        connection.close()
        print("\nâœ… DB ì—°ê²° ì¢…ë£Œ")


# ì‹¤í–‰
if __name__ == "__main__":
    process_db(END_YEAR)  # âœ¨ BOARD_TYPE â†’ END_YEAR
```

```python
# end_year ì»¬ëŸ¼ ì¶”ê°€

# ì „ìê²°ì¬ ëª©ë¡ í¬ë¡¤ëŸ¬ (ê°„í¸ ë²„ì „)
# ìƒì„¸ ì§„ì… ì—†ì´ ëª©ë¡(ê·¸ë¦¬ë“œ)ë§Œ ìˆ˜ì§‘
# ì¤‘ë‹¨ì  ì €ì¥ + ì¬ì‹œì‘ ì§€ì› + JSON + DB ì €ì¥

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from pathlib import Path
from sqlalchemy import create_engine, text
from sqlalchemy.types import BIGINT, VARCHAR
import time
import json
import traceback
import re

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ”§ ì„¤ì • (ì—¬ê¸°ë§Œ ìˆ˜ì •í•˜ì„¸ìš”!)
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

BASE_URL = "http://office.anyfive.com/home/"

# ê¸°ê°„ ì„¤ì •
# 5ë…„ ì´í•˜ ê²½ê³ ì°½ë– ì„œ ë‘ë²ˆ ìª¼ê°œì„œ ì‹¤í–‰í•¨
END_YEAR = 2020  #ì»¨íŠ¸ë¡¤ëŸ¬ - í´ë” ê¸°ê°„ì— ë§ê²Œ ìˆ˜ì • (2025, 2020, 2015, 2010)
# START_DATE = "2016-01-01"  #ì»¨íŠ¸ë¡¤ëŸ¬ - ì‹œì‘ì¼
# END_DATE = "2019-12-31"  #ì»¨íŠ¸ë¡¤ëŸ¬ - ì¢…ë£Œì¼

START_DATE = "2020-01-01"  #ì»¨íŠ¸ë¡¤ëŸ¬ - ì‹œì‘ì¼
END_DATE = "2020-12-31"  #ì»¨íŠ¸ë¡¤ëŸ¬ - ì¢…ë£Œì¼


# DB ì„¤ì •
DB_CONFIG = {
    "user": "root",
    "password": "1234",
    "host": "127.0.0.1",
    "port": 3306,
    "database": "any_approval" #ì»¨íŠ¸ë¡¤ëŸ¬
}

# ì €ì¥ íŒŒì¼ëª…
JSON_FILE = "one_formCrawling_result.json" #ì»¨íŠ¸ë¡¤ëŸ¬
CHECKPOINT_FILE = "one_formCrawling_checkpoint.json" #ì»¨íŠ¸ë¡¤ëŸ¬

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ“¦ ìœ í‹¸ë¦¬í‹° í•¨ìˆ˜
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def norm_space(s):
    """ê³µë°± ì •ê·œí™”"""
    if not s: return ""
    s = s.replace("\xa0", " ")
    s = re.sub(r'[\t\r\n]+', ' ', s)
    s = re.sub(r'\s+', ' ', s)
    return s.strip()

def switch_into_frame_having(driver, selector_css, max_retries=3):
    """selectorê°€ ìˆëŠ” í”„ë ˆì„ìœ¼ë¡œ ì „í™˜"""
    for attempt in range(max_retries):
        driver.switch_to.default_content()
        
        # ë¡œë”© ë§ˆìŠ¤í¬ ëŒ€ê¸°
        try:
            WebDriverWait(driver, 5).until(
                EC.invisibility_of_element_located((By.CSS_SELECTOR, ".k-loading-mask"))
            )
        except:
            pass
        
        time.sleep(1)
        
        # ìµœìƒìœ„ í™•ì¸
        if driver.find_elements(By.CSS_SELECTOR, selector_css):
            print(f"  âœ“ ê·¸ë¦¬ë“œ ë°œê²¬ (ìµœìƒìœ„)")
            return True
        
        # í”„ë ˆì„ ë‚´ë¶€ íƒìƒ‰
        frames = driver.find_elements(By.CSS_SELECTOR, "iframe, frame")
        for idx, fr in enumerate(frames):
            try:
                driver.switch_to.default_content()
                driver.switch_to.frame(fr)
                if driver.find_elements(By.CSS_SELECTOR, selector_css):
                    print(f"  âœ“ ê·¸ë¦¬ë“œ ë°œê²¬ (í”„ë ˆì„ {idx})")
                    return True
            except:
                continue
        
        if attempt < max_retries - 1:
            print(f"  âš ï¸ ì¬ì‹œë„ {attempt + 1}/{max_retries}")
            time.sleep(2)
    
    return False

def wait_grid_stable(driver, timeout=10):
    """ê·¸ë¦¬ë“œ ì•ˆì •í™” ëŒ€ê¸°"""
    try:
        WebDriverWait(driver, timeout).until(
            EC.invisibility_of_element_located((By.CSS_SELECTOR, ".k-loading-mask"))
        )
        WebDriverWait(driver, timeout).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "table[role='grid'] tbody tr"))
        )
        time.sleep(1)
        return True
    except:
        return False

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ’¾ ì²´í¬í¬ì¸íŠ¸ ê´€ë¦¬
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def load_checkpoint():
    """ì²´í¬í¬ì¸íŠ¸ ë¡œë“œ"""
    if Path(CHECKPOINT_FILE).exists():
        with open(CHECKPOINT_FILE, 'r', encoding='utf-8') as f:
            return json.load(f)
    return {"last_page": 0, "total_collected": 0, "last_doc_id": None}

def save_checkpoint(page, total, last_doc_id):
    """ì²´í¬í¬ì¸íŠ¸ ì €ì¥"""
    checkpoint = {
        "last_page": page,
        "total_collected": total,
        "last_doc_id": last_doc_id,
        "timestamp": time.strftime("%Y-%m-%d %H:%M:%S")
    }
    with open(CHECKPOINT_FILE, 'w', encoding='utf-8') as f:
        json.dump(checkpoint, f, ensure_ascii=False, indent=2)
    print(f"  ğŸ’¾ ì²´í¬í¬ì¸íŠ¸ ì €ì¥: í˜ì´ì§€ {page}, ì´ {total}ê±´")

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ’¾ ë°ì´í„° ì €ì¥
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def load_existing_data():
    """ê¸°ì¡´ JSON ë°ì´í„° ë¡œë“œ"""
    if Path(JSON_FILE).exists():
        with open(JSON_FILE, 'r', encoding='utf-8') as f:
            return json.load(f)
    return {}

def save_to_json(data_dict):
    """JSON íŒŒì¼ì— ì €ì¥ (ë³‘í•© ë°©ì‹)"""
    existing = load_existing_data()
    existing.update(data_dict)  # ë³‘í•©!
    
    with open(JSON_FILE, 'w', encoding='utf-8') as f:
        json.dump(existing, f, ensure_ascii=False, indent=2)
    
    print(f"  ğŸ“„ JSON ì €ì¥: {len(existing)}ê±´ (ìƒˆë¡œ ì¶”ê°€ {len(data_dict)}ê±´)")

def save_to_db(data_list, engine, end_year):
    """DBì— ì €ì¥ (UPSERT ë°©ì‹)"""
    if not data_list:
        return
    
    # í…Œì´ë¸” ìƒì„± (ì—†ìœ¼ë©´) 
    create_table_sql = """
    CREATE TABLE IF NOT EXISTS form (
        ë¬¸ì„œID BIGINT PRIMARY KEY,
        ì–‘ì‹ëª… VARCHAR(255),
        ë¬¸ì„œë²ˆí˜¸ VARCHAR(255),
        ì œëª© VARCHAR(500),
        ê¸°ì•ˆì VARCHAR(255),
        ê¸°ì•ˆì¼ VARCHAR(50),
        end_year INT,
        INDEX idx_end_year (end_year)
    ) ENGINE=InnoDB DEFAULT CHARSET=utf8mb4 COLLATE=utf8mb4_unicode_ci;
    """
    
    with engine.connect() as conn:
        conn.execute(text(create_table_sql))
        conn.commit()
    
    # UPSERT (ON DUPLICATE KEY UPDATE)
    upsert_sql = """
    INSERT INTO form 
        (ë¬¸ì„œID, ì–‘ì‹ëª…, ë¬¸ì„œë²ˆí˜¸, ì œëª©, ê¸°ì•ˆì, ê¸°ì•ˆì¼, end_year)
    VALUES 
        (:doc_id, :form_name, :doc_num, :title, :author, :draft_date, :end_year)
    ON DUPLICATE KEY UPDATE
        ì–‘ì‹ëª… = VALUES(ì–‘ì‹ëª…),
        ë¬¸ì„œë²ˆí˜¸ = VALUES(ë¬¸ì„œë²ˆí˜¸),
        ì œëª© = VALUES(ì œëª©),
        ê¸°ì•ˆì = VALUES(ê¸°ì•ˆì),
        ê¸°ì•ˆì¼ = VALUES(ê¸°ì•ˆì¼),
        end_year = VALUES(end_year)
    """
    
    with engine.connect() as conn:
        for item in data_list:
            conn.execute(text(upsert_sql), {
                "doc_id": item["ë¬¸ì„œID"],
                "form_name": item["ì–‘ì‹ëª…"],
                "doc_num": item["ë¬¸ì„œë²ˆí˜¸"],
                "title": item["ì œëª©"],
                "author": item["ê¸°ì•ˆì"],
                "draft_date": item["ê¸°ì•ˆì¼"],
                "end_year": end_year
            })
        conn.commit()
    
    print(f"  ğŸ’¾ DB ì €ì¥: {len(data_list)}ê±´")

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ•·ï¸ ê·¸ë¦¬ë“œ íŒŒì‹±
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def parse_grid_rows(driver, end_year):
    """í˜„ì¬ í˜ì´ì§€ì˜ ê·¸ë¦¬ë“œ íŒŒì‹±"""
    table = driver.find_element(By.CSS_SELECTOR, "table[role='grid']")
    
    # í—¤ë” ë¶„ì„
    th_all = table.find_elements(By.CSS_SELECTOR, "thead th")
    field_index = {}
    for i, th in enumerate(th_all):
        field = (th.get_attribute("data-field") or "").strip()
        if field:
            field_index[field] = i
    
    # í–‰ íŒŒì‹±
    rows = table.find_elements(By.CSS_SELECTOR, "tbody tr[role='row']")
    results = []
    
    for tr in rows:
        tds = tr.find_elements(By.CSS_SELECTOR, "td")
        if not tds:
            continue
        
        # ì²´í¬ë°•ìŠ¤ì—ì„œ ë¬¸ì„œID ì¶”ì¶œ
        cb = tr.find_elements(By.CSS_SELECTOR, "td input[name='aprDoc_grid_ckbx']")
        doc_id = cb[0].get_attribute("value").strip() if cb else ""
        
        def get_td_text(field, fallback_idx=None):
            """field ì´ë¦„ì´ë‚˜ ì¸ë±ìŠ¤ë¡œ td í…ìŠ¤íŠ¸ ì¶”ì¶œ"""
            idx = field_index.get(field)
            if idx is not None and idx < len(tds):
                return norm_space(tds[idx].text or tds[idx].get_attribute("textContent") or "")
            if fallback_idx is not None and fallback_idx < len(tds):
                return norm_space(tds[fallback_idx].text or tds[fallback_idx].get_attribute("textContent") or "")
            return ""
        
        item = {
            "ë¬¸ì„œID": doc_id,
            "ì–‘ì‹ëª…": get_td_text("form_nm", 1),
            "ë¬¸ì„œë²ˆí˜¸": get_td_text("doc_sn", 2),
            "ì œëª©": get_td_text("appr_doc_tit", 3),
            "ê¸°ì•ˆì": get_td_text("kor_nm", 4),
            "ê¸°ì•ˆì¼": get_td_text("drft_dtm", 5),
            "end_year": end_year
        }
        
        results.append(item)
    
    return results

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ”„ í˜ì´ì§•
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def has_next_page(driver):
    """ë‹¤ìŒ í˜ì´ì§€ ì¡´ì¬ ì—¬ë¶€"""
    try:
        next_btns = driver.find_elements(By.CSS_SELECTOR, 
            "a.k-pager-nav[aria-label='ë‹¤ìŒ í˜ì´ì§€ë¡œ ì´ë™']:not(.k-state-disabled)")
        return len(next_btns) > 0
    except:
        return False

def go_to_next_page(driver):
    """ë‹¤ìŒ í˜ì´ì§€ë¡œ ì´ë™"""
    try:
        next_btn = driver.find_element(By.CSS_SELECTOR,
            "a.k-pager-nav[aria-label='ë‹¤ìŒ í˜ì´ì§€ë¡œ ì´ë™']:not(.k-state-disabled)")
        
        driver.execute_script("arguments[0].click();", next_btn)
        time.sleep(2)
        
        # ë¡œë”© ëŒ€ê¸°
        WebDriverWait(driver, 15).until(
            EC.invisibility_of_element_located((By.CSS_SELECTOR, ".k-loading-mask"))
        )
        
        wait_grid_stable(driver)
        return True
    except Exception as e:
        print(f"  âŒ í˜ì´ì§€ ì´ë™ ì‹¤íŒ¨: {e}")
        return False

def go_to_specific_page(driver, target_page):
    """íŠ¹ì • í˜ì´ì§€ë¡œ ì´ë™ (ì¬ì‹œì‘ìš©)"""
    if target_page <= 1:
        return True
    
    print(f"  ğŸ”„ {target_page}í˜ì´ì§€ë¡œ ì´ë™ ì¤‘...")
    
    for page in range(2, target_page + 1):
        if not has_next_page(driver):
            print(f"  âš ï¸ {page}í˜ì´ì§€ì— ë„ë‹¬í•˜ì§€ ëª»í•¨")
            return False
        
        if not go_to_next_page(driver):
            return False
        
        if page % 5 == 0:
            print(f"    ì§„í–‰ ì¤‘... {page}/{target_page}")
    
    print(f"  âœ“ {target_page}í˜ì´ì§€ ë„ì°©")
    return True

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸš€ ë©”ì¸ í¬ë¡¤ëŸ¬
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def main():
    print("=" * 70)
    print("ğŸ•·ï¸  ì „ìê²°ì¬ ëª©ë¡ í¬ë¡¤ëŸ¬ ì‹œì‘")
    print(f"ğŸ“… ê¸°ê°„: {START_DATE} ~ {END_DATE} (endYear={END_YEAR})")
    print("=" * 70)
    
    # ì²´í¬í¬ì¸íŠ¸ ë¡œë“œ
    checkpoint = load_checkpoint()
    start_page = checkpoint["last_page"] + 1
    total_collected = checkpoint["total_collected"]
    
    if start_page > 1:
        print(f"\nğŸ”„ ì¬ì‹œì‘ ëª¨ë“œ")
        print(f"  ë§ˆì§€ë§‰ ì™„ë£Œ í˜ì´ì§€: {checkpoint['last_page']}")
        print(f"  ê¸°ì¡´ ìˆ˜ì§‘ëŸ‰: {total_collected}ê±´")
        print(f"  {start_page}í˜ì´ì§€ë¶€í„° ì´ì–´ì„œ ì‹œì‘í•©ë‹ˆë‹¤.\n")
    else:
        print(f"\nğŸ†• ìƒˆë¡œ ì‹œì‘\n")
    
    # DB ì—°ê²°
    try:
        engine = create_engine(
            f"mysql+pymysql://{DB_CONFIG['user']}:{DB_CONFIG['password']}@"
            f"{DB_CONFIG['host']}:{DB_CONFIG['port']}/{DB_CONFIG['database']}"
            f"?charset=utf8mb4"
        )
        print(f"âœ… DB ì—°ê²° ì„±ê³µ: {DB_CONFIG['host']}:{DB_CONFIG['port']}/{DB_CONFIG['database']}\n")
    except Exception as e:
        print(f"âŒ DB ì—°ê²° ì‹¤íŒ¨: {e}")
        return
    
    # ë¸Œë¼ìš°ì € ì‹œì‘
    opts = webdriver.ChromeOptions()
    opts.add_argument("--start-maximized")
    opts.add_argument("--disable-popup-blocking")
    opts.add_argument("--disable-blink-features=AutomationControlled")
    opts.add_experimental_option("excludeSwitches", ["enable-automation"])
    
    driver = webdriver.Chrome(service=Service(), options=opts)
    
    try:
        # ë¡œê·¸ì¸
        driver.get(BASE_URL)
        time.sleep(1)
        
        print("\n" + "=" * 70)
        print("ğŸ“‹ ìˆ˜ë™ ì‘ì—…ì´ í•„ìš”í•©ë‹ˆë‹¤!")
        print("=" * 70)
        print("ë‹¤ìŒ ë‹¨ê³„ë¥¼ ì™„ë£Œí•œ í›„ Enterë¥¼ ëˆŒëŸ¬ì£¼ì„¸ìš”:\n")
        print("  1ï¸âƒ£  ë¡œê·¸ì¸")
        print("  2ï¸âƒ£  ì „ìê²°ì¬ ë©”ë‰´ í´ë¦­")
        print("  3ï¸âƒ£  ê²°ì¬ ë¬¸ì„œê´€ë¦¬ í´ë¦­")
        print(f"  4ï¸âƒ£  ìƒíƒœ = ì™„ë£Œ ì„ íƒ")
        print(f"  5ï¸âƒ£  ì‹œì‘ì¼ = {START_DATE}")
        print(f"  6ï¸âƒ£  ì¢…ë£Œì¼ = {END_DATE}")
        print("  7ï¸âƒ£  ì¡°íšŒ ë²„íŠ¼ í´ë¦­ â†’ ê·¸ë¦¬ë“œ ë‚˜íƒ€ë‚¨ í™•ì¸")
        print("\nğŸ’¡ ê·¸ë¦¬ë“œê°€ í™”ë©´ì— ë³´ì´ë©´ Enterë¥¼ ëˆ„ë¥´ì„¸ìš”!")
        print("=" * 70)
        input("\nâœ‹ ì¤€ë¹„ë˜ë©´ Enter... ")
        
        time.sleep(1)
        
        # ê·¸ë¦¬ë“œë¡œ ì „í™˜
        if not switch_into_frame_having(driver, "table[role='grid']"):
            print("âŒ ê·¸ë¦¬ë“œë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤")
            return
        
        wait_grid_stable(driver)
        
        # ì¬ì‹œì‘ ì‹œ íŠ¹ì • í˜ì´ì§€ë¡œ ì´ë™
        if start_page > 1:
            if not go_to_specific_page(driver, start_page):
                print("âŒ ì‹œì‘ í˜ì´ì§€ë¡œ ì´ë™ ì‹¤íŒ¨")
                return
        
        # í˜ì´ì§€ë³„ ìˆ˜ì§‘
        current_page = start_page
        
        while True:
            print(f"\n{'=' * 70}")
            print(f"ğŸ“„ í˜ì´ì§€ {current_page} ìˆ˜ì§‘ ì¤‘...")
            print(f"{'=' * 70}")
            
            try:
                # ê·¸ë¦¬ë“œ ì¬í™•ì¸
                if not switch_into_frame_having(driver, "table[role='grid']"):
                    print("âš ï¸ ê·¸ë¦¬ë“œ ì¬íƒìƒ‰ ì‹¤íŒ¨")
                    break
                
                wait_grid_stable(driver)
                
                # íŒŒì‹±
                page_data = parse_grid_rows(driver, END_YEAR)
                
                if not page_data:
                    print("âš ï¸ ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤")
                    break
                
                print(f"  âœ“ {len(page_data)}ê±´ íŒŒì‹± ì™„ë£Œ")
                
                # JSON ì €ì¥ (ë³‘í•©)
                json_dict = {item["ë¬¸ì„œID"]: {
                    "ì–‘ì‹ëª…": item["ì–‘ì‹ëª…"],
                    "ë¬¸ì„œë²ˆí˜¸": item["ë¬¸ì„œë²ˆí˜¸"],
                    "ì œëª©": item["ì œëª©"],
                    "ê¸°ì•ˆì": item["ê¸°ì•ˆì"],
                    "ê¸°ì•ˆì¼": item["ê¸°ì•ˆì¼"],
                    "end_year": item["end_year"]
                } for item in page_data}
                
                save_to_json(json_dict)
                
                # DB ì €ì¥ (UPSERT)
                save_to_db(page_data, engine, END_YEAR)
                
                # ì²´í¬í¬ì¸íŠ¸ ì €ì¥
                total_collected += len(page_data)
                last_doc_id = page_data[-1]["ë¬¸ì„œID"] if page_data else None
                save_checkpoint(current_page, total_collected, last_doc_id)
                
                print(f"  âœ… í˜ì´ì§€ {current_page} ì™„ë£Œ (ëˆ„ì : {total_collected}ê±´)")
                
                # ë‹¤ìŒ í˜ì´ì§€
                if has_next_page(driver):
                    if go_to_next_page(driver):
                        current_page += 1
                        time.sleep(1)
                    else:
                        break
                else:
                    print(f"\nğŸ‰ ë§ˆì§€ë§‰ í˜ì´ì§€ ë„ë‹¬!")
                    break
                
            except Exception as e:
                print(f"âŒ í˜ì´ì§€ {current_page} ì²˜ë¦¬ ì‹¤íŒ¨: {e}")
                traceback.print_exc()
                break
        
        print(f"\n{'=' * 70}")
        print(f"âœ… ìˆ˜ì§‘ ì™„ë£Œ!")
        print(f"{'=' * 70}")
        print(f"ğŸ“Š ì´ ìˆ˜ì§‘: {total_collected}ê±´")
        print(f"ğŸ“„ JSON: {JSON_FILE}")
        print(f"ğŸ’¾ DB: form í…Œì´ë¸” (endYear={END_YEAR})")
        print(f"ğŸ“ ì²´í¬í¬ì¸íŠ¸: {CHECKPOINT_FILE}")
        print(f"{'=' * 70}\n")
        
    except Exception as e:
        print(f"\nâŒ ì˜¤ë¥˜ ë°œìƒ: {e}")
        traceback.print_exc()
    
    finally:
        input("\nâœ‹ ë¸Œë¼ìš°ì €ë¥¼ ë‹«ìœ¼ë ¤ë©´ Enterë¥¼ ëˆ„ë¥´ì„¸ìš”...")
        driver.quit()

if __name__ == "__main__":
    main()
```

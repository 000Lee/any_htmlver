```python
# references가 존재하는 문서들의 sourceId 숫자를 뽑는 코드
import json
import re

result = []

# cmds 파일 읽기 (정확한 파일명 사용)
with open('documents_2025_02.cmds', 'r', encoding='utf-8') as f:
    for line in f:
        line = line.strip()
        
        # addDocument로 시작하는 줄만 처리
        if line.startswith('addDocument '):
            # "addDocument " 이후의 JSON 부분만 추출
            json_str = line.replace('addDocument ', '', 1)
            
            try:
                data = json.loads(json_str)
                
                # references가 존재하고 비어있지 않은 경우
                if 'references' in data and data['references']:
                    # sourceId에서 숫자만 추출
                    source_id = data['sourceId']  # 예: "doc_18876965_02"
                    match = re.search(r'doc_(\d+)_\d+', source_id)
                    if match:
                        number = match.group(1)  # "18876965"
                        result.append(number)
            except json.JSONDecodeError:
                # JSON 파싱 에러 시 스킵
                continue

print(f"추출된 숫자 개수: {len(result)}")
print(result)
```

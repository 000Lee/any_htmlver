```python
import json

# 1. 첫 번째 JSON 파일 읽기 (메타데이터 배열) 
#컨트롤러
with open('two_htmltojsondb_convert.json', 'r', encoding='utf-8') as f:
    metadata_list = json.load(f)

# 2. 두 번째 JSON 파일 읽기 (HTML 내용 객체)
#컨트롤러
with open('two_htmltodocbodyconvert_result.json', 'r', encoding='utf-8') as f:
    html_dict = json.load(f)

# 3. sourceId를 기준으로 docBody에 HTML 내용 삽입
for doc in metadata_list:
    source_id = doc.get('sourceId')
    
    # sourceId가 html_dict에 있으면 docBody에 삽입
    if source_id and source_id in html_dict:
        doc['docBody'] = html_dict[source_id]
        print(f"✓ {source_id}: docBody 삽입 완료")
    else:
        print(f"✗ {source_id}: HTML 내용을 찾을 수 없음")

# 4. 결과를 새 파일로 저장
#컨트롤러
with open('three_docbody_insert.json', 'w', encoding='utf-8') as f:
    json.dump(metadata_list, f, ensure_ascii=False, indent=2)

print("\n완료! three_docbody_insert.json 파일에 저장되었습니다.")
```

```python
# styleíŒŒì¼ apr.doc.print.jstl.cssë§Œ ì‚¬ìš©
# <!DOCTYPE html><html lang="en"><head><body>íƒœê·¸ ì œê±°
# .content_box{min-width:850->819pxë¡œ ë³€ê²½, width:99.9% ì‚­ì œ
# ul,ol{list-style:none !important; padding-left: 0 !important; margin-left: 0 !important;} ìœ¼ë¡œ ë³€ê²½
# .content_box{width:97.7% !important;}ì‚­ì œ

#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
 [ê¸°ëŠ¥]
 HTML íŒŒì¼ì— CSSë¥¼ ì¸ë¼ì¸ ì‚½ì…í•˜ê³  ì••ì¶•í•˜ì—¬ DBì— ì €ì¥

 [ì²˜ë¦¬ ìˆœì„œ]
 1. CSS íŒŒì¼ ì½ê¸° (apr.doc.print.jstl.cssë§Œ ì‚¬ìš©)
 2. CSS ê·œì¹™ ìˆ˜ì •
    - .content_box: min-width 850pxâ†’819px, width:99.9% ì‚­ì œ
    - ul,ol: list-style:none !important + padding/margin 0
 3. HTMLì˜ <link> íƒœê·¸ë¥¼ <style> íƒœê·¸ë¡œ êµì²´ (CSS ì¸ë¼ì¸í™”)
 4. ë¶ˆí•„ìš”í•œ ì„¹ì…˜ ì‚­ì œ
    - ì²¨ë¶€íŒŒì¼, ì°¸ì¡°ë¬¸ì„œ, ê²°ì¬ì˜ê²¬, ê²°ì¬ëŒ“ê¸€ ì˜ì—­
 5. ì»¤ìŠ¤í…€ CSS ì¶”ê°€ (.content, .content table ë„ˆë¹„ ì¡°ì •)
 6. <style> íƒœê·¸ë¥¼ </body> ì§ì „ìœ¼ë¡œ ì´ë™
 7. HTML wrapper íƒœê·¸ ì œê±°
    - <!DOCTYPE html>, <html>, <head>, <body> ë“±
 8. HTML/CSS ì••ì¶• (í•œ ì¤„ë¡œ minify)
 9. DBì˜ doc_body ì»¬ëŸ¼ì— UPDATE

 [ì¶œë ¥]
 - MariaDB: documents í…Œì´ë¸”ì˜ doc_body ì»¬ëŸ¼ ì—…ë°ì´íŠ¸
 - ì‹¤íŒ¨ ì‹œ: failed_updates.txt íŒŒì¼ ìƒì„±

 [ì„¤ì • (#ì—¬ê¸°ë¥¼ ìˆ˜ì •í•˜ì„¸ìš”)]
 - base_path: HTML í´ë” ê²½ë¡œ
 - end_year: ê¸°ê°„ êµ¬ë¶„ìš© ì—°ë„
 - db_config: DB ì ‘ì† ì •ë³´ (host, user, password, database)

 [ì˜ì¡´ì„±]
 - beautifulsoup4
 - pymysql
-> pip install beautifulsoup4 pymysql lxml
"""
import json
import re
from pathlib import Path
from bs4 import BeautifulSoup
import pymysql

def minify_css(css):
    """CSSë¥¼ í•œ ì¤„ë¡œ ì••ì¶• (ë Œë”ë§ ê²°ê³¼ëŠ” ë™ì¼)"""
    # CSS ì£¼ì„ ì œê±°
    css = re.sub(r'/\*.*?\*/', '', css, flags=re.DOTALL)
    # ì¤„ë°”ê¿ˆ ì œê±°
    css = css.replace('\n', '').replace('\r', '')
    # ì—°ì†ëœ ê³µë°±ì„ í•˜ë‚˜ë¡œ
    css = re.sub(r'\s+', ' ', css)
    # { } : ; ì•ë’¤ ê³µë°± ì œê±°
    css = re.sub(r'\s*{\s*', '{', css)
    css = re.sub(r'\s*}\s*', '}', css)
    css = re.sub(r'\s*:\s*', ':', css)
    css = re.sub(r'\s*;\s*', ';', css)
    css = re.sub(r'\s*,\s*', ',', css)
    return css.strip()

def minify_html(html):
    """
    HTMLì„ ì™„ì „íˆ í•œ ì¤„ë¡œ ì••ì¶• (CSS í¬í•¨)
    - <pre>, <textarea> íƒœê·¸ ë‚´ë¶€ë§Œ ë³´ì¡´
    - CSS, Scriptë„ ëª¨ë‘ í•œ ì¤„ë¡œ ì••ì¶•
    """
    # 1. <style> íƒœê·¸ ë‚´ë¶€ CSSë¥¼ ì••ì¶•
    def minify_style_tag(match):
        tag_open = match.group(1)  # <style...>
        css_content = match.group(2)
        minified_css = minify_css(css_content)
        return f"{tag_open}{minified_css}</style>"
    
    html = re.sub(r'(<style[^>]*>)(.*?)</style>', minify_style_tag, html, flags=re.DOTALL | re.IGNORECASE)
    
    # 2. <script> íƒœê·¸ ë‚´ë¶€ë¥¼ í•œ ì¤„ë¡œ ì••ì¶•
    def minify_script_tag(match):
        tag_open = match.group(1)
        script_content = match.group(2)
        # ìŠ¤í¬ë¦½íŠ¸ëŠ” ê°„ë‹¨íˆ ì¤„ë°”ê¿ˆê³¼ ì—°ì† ê³µë°±ë§Œ ì œê±°
        minified = script_content.replace('\n', '').replace('\r', '')
        minified = re.sub(r'\s+', ' ', minified)
        return f"{tag_open}{minified}</script>"
    
    html = re.sub(r'(<script[^>]*>)(.*?)</script>', minify_script_tag, html, flags=re.DOTALL | re.IGNORECASE)
    
    # 3. <pre>, <textarea> íƒœê·¸ ë‚´ë¶€ë¥¼ ì„ì‹œë¡œ ë³´í˜¸
    pre_blocks = []
    def save_pre(match):
        pre_blocks.append(match.group(0))
        return f"___PRE_BLOCK_{len(pre_blocks)-1}___"
    
    html = re.sub(r'<pre[^>]*>.*?</pre>', save_pre, html, flags=re.DOTALL | re.IGNORECASE)
    html = re.sub(r'<textarea[^>]*>.*?</textarea>', save_pre, html, flags=re.DOTALL | re.IGNORECASE)
    
    # 4. HTML ì£¼ì„ ì œê±°
    html = re.sub(r'<!--.*?-->', '', html, flags=re.DOTALL)
    
    # 5. íƒœê·¸ ì‚¬ì´ì˜ ë¶ˆí•„ìš”í•œ ê³µë°± ì œê±°
    html = re.sub(r'>\s+<', '><', html)
    
    # 6. ì—°ì†ëœ ê³µë°±ì„ í•˜ë‚˜ë¡œ
    html = re.sub(r'\s+', ' ', html)
    
    # 7. ì¤„ë°”ê¿ˆ ì œê±°
    html = html.replace('\n', '').replace('\r', '')
    
    # 8. ë³´í˜¸í–ˆë˜ pre/textarea ë¸”ë¡ë“¤ ë³µì›
    for i, block in enumerate(pre_blocks):
        html = html.replace(f"___PRE_BLOCK_{i}___", block)
    
    return html.strip()

def extract_doc_id(filename):
    """íŒŒì¼ëª…ì—ì„œ ë¬¸ì„œID ì¶”ì¶œ (ì˜ˆ: 20151224_ì œëª©_2009465.html -> 2009465)"""
    # .html ì œê±°
    name_without_ext = filename.replace('.html', '')
    # ë§ˆì§€ë§‰ ì–¸ë”ìŠ¤ì½”ì–´ ë’¤ì˜ ìˆ«ì ë¶€ë¶„ì´ ë¬¸ì„œID
    parts = name_without_ext.split('_')
    if len(parts) >= 2:
        # ë§ˆì§€ë§‰ ë¶€ë¶„ì´ ìˆ«ìì¸ì§€ í™•ì¸
        last_part = parts[-1]
        if last_part.isdigit():
            return last_part
    return None

def modify_specific_css_rules(css_content):
    """íŠ¹ì • CSS ì„ íƒìì˜ ì†ì„±ë§Œ ìˆ˜ì •"""
    
    # 1. .content_box ê·œì¹™ ìˆ˜ì •
    # .content_box{ ... } ë¸”ë¡ì„ ì°¾ì•„ì„œ ë‚´ë¶€ ì†ì„± ìˆ˜ì •
    def modify_content_box(match):
        full_block = match.group(0)
        # min-width:850px â†’ min-width:819px
        full_block = re.sub(r'min-width\s*:\s*850px', 'min-width:819px', full_block)
        # width:99.9%; ì œê±°
        full_block = re.sub(r'width\s*:\s*99\.9%\s*;?', '', full_block)
        return full_block
    
    # .content_box{...} ë¸”ë¡ ì°¾ê¸° (ì¤‘ê´„í˜¸ ì•ˆì˜ ë‚´ìš© í¬í•¨)
    css_content = re.sub(
        r'\.content_box\s*\{[^}]*\}',
        modify_content_box,
        css_content,
        flags=re.DOTALL
    )
    
    # 2. ul,ol ê·œì¹™ ìˆ˜ì •
    # ul,ol{list-style:none} â†’ ul,ol{list-style:none !important; padding-left: 0 !important; margin-left: 0 !important;}
    def modify_ul_ol(match):
        # ê¸°ì¡´ ë¸”ë¡ì„ ìƒˆë¡œìš´ ê·œì¹™ìœ¼ë¡œ êµì²´
        return 'ul,ol{list-style:none !important; padding-left: 0 !important; margin-left: 0 !important;}'
    
    # ul,ol{...} ë¸”ë¡ ì°¾ê¸°
    css_content = re.sub(
        r'ul\s*,\s*ol\s*\{[^}]*\}',
        modify_ul_ol,
        css_content,
        flags=re.DOTALL
    )
    
    return css_content

def remove_html_wrapper_tags(html_content):
    """HTML wrapper íƒœê·¸ë§Œ ì œê±° (ë‚´ìš©ì€ ìœ ì§€)"""
    
    # ì •í™•í•œ íƒœê·¸ë§Œ í•œ ì¤„ì”© ì œê±°
    html_content = re.sub(r'<!DOCTYPE html>\s*', '', html_content, flags=re.IGNORECASE)
    html_content = re.sub(r'<html[^>]*>\s*', '', html_content, flags=re.IGNORECASE)
    html_content = re.sub(r'<head>\s*', '', html_content, flags=re.IGNORECASE)
    html_content = re.sub(r'</head>\s*', '', html_content, flags=re.IGNORECASE)
    html_content = re.sub(r'<body>\s*', '', html_content, flags=re.IGNORECASE)
    html_content = re.sub(r'</body>\s*', '', html_content, flags=re.IGNORECASE)
    html_content = re.sub(r'</html>\s*', '', html_content, flags=re.IGNORECASE)
    
    return html_content

def read_css_files(css_folder):
    """CSS íŒŒì¼ 1ê°œë¥¼ ì½ì–´ì„œ ë”•ì…”ë„ˆë¦¬ë¡œ ë°˜í™˜"""
    css_dict = {}
    css_files = [
        'apr.doc.print.jstl.css'  # 1ê°œë§Œ ì‚¬ìš©
    ]
    
    for css_file in css_files:
        css_path = css_folder / css_file
        if css_path.exists():
            with open(css_path, 'r', encoding='utf-8', errors='ignore') as f:
                css_content = f.read()
            
            # CSS ê·œì¹™ ìˆ˜ì •
            css_content = modify_specific_css_rules(css_content)
            css_dict[css_file] = css_content
            
            print(f"âœ“ CSS íŒŒì¼ ì½ê¸° ë° ê·œì¹™ ìˆ˜ì • ì™„ë£Œ: {css_file}")
        else:
            print(f"âš ï¸  CSS íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {css_file}")
    
    return css_dict

def insert_css_inline(html_content, css_dict):
    """HTMLì˜ <link> íƒœê·¸ë¥¼ <style> íƒœê·¸ë¡œ êµì²´"""
    soup = BeautifulSoup(html_content, 'html.parser')
    
    # ëª¨ë“  <link> íƒœê·¸ ì°¾ê¸°
    link_tags = soup.find_all('link', rel='stylesheet')
    
    for link_tag in link_tags:
        href = link_tag.get('href', '')
        
        # CSS íŒŒì¼ëª… ì¶”ì¶œ
        css_filename = None
        if 'apr.doc.print.jstl.css' in href:
            css_filename = 'apr.doc.print.jstl.css'
        
        # í•´ë‹¹ CSSê°€ ìˆìœ¼ë©´ <style> íƒœê·¸ë¡œ êµì²´
        if css_filename and css_filename in css_dict:
            style_tag = soup.new_tag('style')
            style_tag.string = css_dict[css_filename]
            link_tag.replace_with(style_tag)
    
    return str(soup)

def remove_specific_sections(html_content):
    """ì²¨ë¶€íŒŒì¼, ì°¸ì¡°ë¬¸ì„œ, ê²°ì¬ì˜ê²¬, ê²°ì¬ëŒ“ê¸€ ì„¹ì…˜ ì‚­ì œ"""
    soup = BeautifulSoup(html_content, 'html.parser')
    
    # 1. ì²¨ë¶€íŒŒì¼ ì˜ì—­ ì‚­ì œ (id="attachFileArea"ì¸ div)
    for div in soup.find_all('div', id='attachFileArea'):
        div.decompose()
    
    # 2. ì²¨ë¶€íŒŒì¼ ì˜ì—­ ì‚­ì œ (captionì´ "íŒŒì¼,ë¬¸ì„œ"ì¸ ê²½ìš°)
    for div in soup.find_all('div', class_='table05_box'):
        caption = div.find('caption')
        if caption and 'íŒŒì¼' in caption.get_text(strip=True):
            div.decompose()
    
    # 3. ì²¨ë¶€íŒŒì¼ í–‰ ì‚­ì œ (thì— "ì²¨ë¶€íŒŒì¼" í…ìŠ¤íŠ¸ê°€ ìˆëŠ” tr)
    for th in soup.find_all('th'):
        if 'ì²¨ë¶€íŒŒì¼' in th.get_text(strip=True):
            tr = th.find_parent('tr')
            if tr:
                tr.decompose()
    
    # 4. ì°¸ì¡°ë¬¸ì„œ í–‰ ì‚­ì œ (thì— "ì°¸ì¡°ë¬¸ì„œ" í…ìŠ¤íŠ¸ê°€ ìˆëŠ” tr)
    for th in soup.find_all('th', scope='row'):
        if th.get_text(strip=True) == 'ì°¸ì¡°ë¬¸ì„œ':
            tr = th.find_parent('tr')
            if tr:
                tr.decompose()
    
    # 5. ê²°ì¬ì˜ê²¬ ì˜ì—­ ì‚­ì œ
    for div in soup.find_all('div', class_='table04_box'):
        caption = div.find('caption')
        if caption and 'ê²°ì¬ì˜ê²¬' in caption.get_text(strip=True):
            div.decompose()
    
    # 6. ê²°ì¬ëŒ“ê¸€ ì˜ì—­ ì‚­ì œ
    for div in soup.find_all('div', class_='table04_box'):
        caption = div.find('caption')
        if caption and 'ê²°ì¬ëŒ“ê¸€' in caption.get_text(strip=True):
            div.decompose()
    
    return str(soup)

def add_custom_css_to_styles(html_content):
    """ëª¨ë“  <style> íƒœê·¸ ë§¨ ì•ì— ì»¤ìŠ¤í…€ CSS ì¶”ê°€"""
    soup = BeautifulSoup(html_content, 'html.parser')
    
    custom_css = '.content{width: 80% !important;} .content table {width: 100% !important;}'
    
    # ëª¨ë“  <style> íƒœê·¸ ì°¾ê¸°
    style_tags = soup.find_all('style')
    
    for style_tag in style_tags:
        # ê¸°ì¡´ ë‚´ìš© ì•ì— ì»¤ìŠ¤í…€ CSS ì¶”ê°€
        original_content = style_tag.string if style_tag.string else ''
        style_tag.string = custom_css + original_content
    
    return str(soup)

def move_styles_before_body_close(html_content):
    """ëª¨ë“  <style> íƒœê·¸ë¥¼ </body> íƒœê·¸ ì§ì „ìœ¼ë¡œ ì´ë™"""
    soup = BeautifulSoup(html_content, 'html.parser')
    
    # ëª¨ë“  <style> íƒœê·¸ ì°¾ê¸°
    style_tags = soup.find_all('style')
    
    # body íƒœê·¸ ì°¾ê¸°
    body = soup.find('body')
    
    if body and style_tags:
        # ê¸°ì¡´ ìœ„ì¹˜ì—ì„œ style íƒœê·¸ë“¤ì„ ì œê±°í•˜ê³  ìˆ˜ì§‘
        for style_tag in style_tags:
            style_tag.extract()
        
        # body íƒœê·¸ ëì— ì¶”ê°€
        for style_tag in style_tags:
            body.append(style_tag)
    
    return str(soup)

def check_and_fix_column_type(db_config):
    """doc_body ì»¬ëŸ¼ íƒ€ì… í™•ì¸ ë° í•„ìš”ì‹œ ë³€ê²½"""
    conn = None
    cursor = None
    
    try:
        conn = pymysql.connect(
            host=db_config['host'],
            user=db_config['user'],
            password=db_config['password'],
            database=db_config['database'],
            charset='utf8mb4'
        )
        cursor = conn.cursor()
        
        # ì»¬ëŸ¼ íƒ€ì… í™•ì¸
        cursor.execute("SHOW COLUMNS FROM documents LIKE 'doc_body'")
        result = cursor.fetchone()
        
        if result:
            column_type = result[1].decode('utf-8') if isinstance(result[1], bytes) else result[1]
            print(f"\n=== doc_body ì»¬ëŸ¼ íƒ€ì… í™•ì¸ ===")
            print(f"í˜„ì¬ íƒ€ì…: {column_type}")
            
            # TEXT íƒ€ì…ì´ë©´ MEDIUMTEXTë¡œ ë³€ê²½
            if column_type.upper() in ['TEXT', 'VARCHAR']:
                print(f"âš ï¸  {column_type}ëŠ” í¬ê¸°ê°€ ë¶€ì¡±í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.")
                print(f"ğŸ”„ MEDIUMTEXTë¡œ ë³€ê²½ ì¤‘...")
                
                cursor.execute("ALTER TABLE documents MODIFY COLUMN doc_body MEDIUMTEXT")
                conn.commit()
                print(f"âœ… doc_body ì»¬ëŸ¼ì„ MEDIUMTEXTë¡œ ë³€ê²½ ì™„ë£Œ!")
            elif column_type.upper() in ['MEDIUMTEXT', 'LONGTEXT']:
                print(f"âœ… ì»¬ëŸ¼ íƒ€ì…ì´ ì¶©ë¶„í•©ë‹ˆë‹¤.")
            else:
                print(f"âš ï¸  ì˜ˆìƒì¹˜ ëª»í•œ íƒ€ì…ì…ë‹ˆë‹¤: {column_type}")
        
    except Exception as e:
        print(f"âŒ ì»¬ëŸ¼ íƒ€ì… í™•ì¸/ë³€ê²½ ì¤‘ ì˜¤ë¥˜: {e}")
    
    finally:
        if cursor:
            cursor.close()
        if conn:
            conn.close()

def update_db_docbody(processed_data, db_config, end_year):
    """ì²˜ë¦¬ëœ HTMLì„ DBì˜ doc_body ì»¬ëŸ¼ì— ì—…ë°ì´íŠ¸"""
    conn = None
    cursor = None
    
    try:
        print("\n=== DB ì—°ê²° ì‹œì‘ ===")
        print(f"Host: {db_config['host']}, Database: {db_config['database']}")
        
        # DB ì—°ê²°
        conn = pymysql.connect(
            host=db_config['host'],
            user=db_config['user'],
            password=db_config['password'],
            database=db_config['database'],
            charset='utf8mb4'
        )
        print("âœ“ DB ì—°ê²° ì„±ê³µ")
        
        cursor = conn.cursor()
        
        # ì—…ë°ì´íŠ¸ ì¿¼ë¦¬
        update_sql = """
        UPDATE documents 
        SET doc_body = %s 
        WHERE source_id = %s AND end_year = %s
        """
        
        success_count = 0
        fail_count = 0
        failed_items = []  # ì‹¤íŒ¨í•œ í•­ëª© ì €ì¥
        
        print(f"\n=== DB ì—…ë°ì´íŠ¸ ì‹œì‘ ({len(processed_data)}ê±´) ===")
        
        for idx, (doc_id, html_content) in enumerate(processed_data.items(), 1):
            try:
                html_size = len(html_content.encode('utf-8'))
                cursor.execute(update_sql, (html_content, doc_id, end_year))
                success_count += 1
                
                # 100ê±´ë§ˆë‹¤ ì¶œë ¥
                if idx % 100 == 0 or idx == len(processed_data):
                    print(f"  ì§„í–‰: {idx}/{len(processed_data)} (ì„±ê³µ: {success_count}, ì‹¤íŒ¨: {fail_count})")
                    
            except Exception as e:
                fail_count += 1
                html_size = len(html_content.encode('utf-8'))
                failed_items.append({
                    'doc_id': doc_id,
                    'size': html_size,
                    'error': str(e)
                })
                
                if fail_count <= 10:
                    print(f"  [ì˜¤ë¥˜ {fail_count}] source_id: {doc_id}, í¬ê¸°: {html_size:,} bytes - {str(e)[:80]}")
        
        # ì»¤ë°‹
        print("\n=== ì»¤ë°‹ ì‹œì‘ ===")
        conn.commit()
        print(f"âœ“ ì»¤ë°‹ ì™„ë£Œ")
        
        # ê²°ê³¼ ìš”ì•½
        print(f"\n=== DB ì—…ë°ì´íŠ¸ ê²°ê³¼ ===")
        print(f"âœ“ ì„±ê³µ: {success_count}ê±´")
        print(f"âœ— ì‹¤íŒ¨: {fail_count}ê±´")
        
        # ì‹¤íŒ¨í•œ í•­ëª© ìƒì„¸ ì •ë³´
        if failed_items:
            print(f"\n=== ì‹¤íŒ¨ í•­ëª© ë¶„ì„ ===")
            sizes = [item['size'] for item in failed_items]
            print(f"ì‹¤íŒ¨í•œ í•­ëª© í‰ê·  í¬ê¸°: {sum(sizes)/len(sizes):,.0f} bytes")
            print(f"ì‹¤íŒ¨í•œ í•­ëª© ìµœëŒ€ í¬ê¸°: {max(sizes):,} bytes")
            print(f"ì‹¤íŒ¨í•œ í•­ëª© ìµœì†Œ í¬ê¸°: {min(sizes):,} bytes")
            
            # ì‹¤íŒ¨ í•­ëª©ì„ íŒŒì¼ë¡œ ì €ì¥
            with open('failed_updates.txt', 'w', encoding='utf-8') as f:
                for item in failed_items:
                    f.write(f"source_id: {item['doc_id']}, size: {item['size']:,} bytes, error: {item['error']}\n")
            print(f"âœ“ ì‹¤íŒ¨ í•­ëª© ìƒì„¸ ì •ë³´ë¥¼ 'failed_updates.txt'ì— ì €ì¥í–ˆìŠµë‹ˆë‹¤.")
        
    except pymysql.Error as e:
        print(f"\nâŒ DB ì˜¤ë¥˜ ë°œìƒ:")
        print(f"  Error Code: {e.args[0]}")
        print(f"  Error Message: {e.args[1]}")
        
    except Exception as e:
        print(f"\nâŒ ì¼ë°˜ ì˜¤ë¥˜ ë°œìƒ: {e}")
        
    finally:
        if cursor:
            cursor.close()
        if conn:
            conn.close()
            print("\nâœ“ DB ì—°ê²° ì¢…ë£Œ")

def main():
    # 1. ê²½ë¡œ ì„¤ì • (ì‚¬ìš©ì í™˜ê²½ì— ë§ê²Œ ìˆ˜ì • í•„ìš”)
    
    base_path = Path("C:/Users/LEEJUHWAN/Downloads/2021-01-01~2025-10-31/html") #ì—¬ê¸°ë¥¼ ìˆ˜ì •í•˜ì„¸ìš”
    end_year = 2025  #ì—¬ê¸°ë¥¼ ìˆ˜ì •í•˜ì„¸ìš” - í´ë” ê¸°ê°„ì— ë§ê²Œ ìˆ˜ì •
    approval_folder = base_path / "ê²°ì¬"
    css_folder = base_path / "resource" / "css"
    
    print(f"ğŸ“‚ ê¸°ë³¸ ê²½ë¡œ: {base_path}")
    print(f"ğŸ“‚ ê²°ì¬ í´ë”: {approval_folder}")
    print(f"ğŸ“‚ CSS í´ë”: {css_folder}\n")
    
    # ê²½ë¡œ ì¡´ì¬ í™•ì¸
    if not approval_folder.exists():
        print(f"âŒ ê²°ì¬ í´ë”ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {approval_folder}")
        print(f"ğŸ’¡ base_path ë³€ìˆ˜ë¥¼ ì‹¤ì œ ê²½ë¡œë¡œ ìˆ˜ì •í•´ì£¼ì„¸ìš”!")
        return
    
    if not css_folder.exists():
        print(f"âŒ CSS í´ë”ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {css_folder}")
        return
    
    # DB ì„¤ì •
    db_config = {
        'host': 'localhost',
        'user': 'root',
        'password': '1234',
        'database': 'any_approval' #ì—¬ê¸°ë¥¼ ìˆ˜ì •í•˜ì„¸ìš” - í†µí•© DBëª…
    }
    
    # DB ì»¬ëŸ¼ íƒ€ì… í™•ì¸ ë° í•„ìš”ì‹œ ë³€ê²½
    check_and_fix_column_type(db_config)
    
    # 2. CSS íŒŒì¼ ì½ê¸° ë° ê·œì¹™ ìˆ˜ì •
    print(f"\nğŸ”„ CSS íŒŒì¼ ì½ê¸° ë° ê·œì¹™ ìˆ˜ì • ì¤‘...\n")
    css_dict = read_css_files(css_folder)
    
    if not css_dict:
        print(f"âŒ CSS íŒŒì¼ì„ ì½ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        return
    
    print(f"\nâœ… {len(css_dict)}ê°œ CSS íŒŒì¼ ë¡œë“œ ë° ìˆ˜ì • ì™„ë£Œ\n")
    
    # 3. ì—°ë„ë³„ í´ë”ì—ì„œ HTML íŒŒì¼ ìˆ˜ì§‘
    print(f"ğŸ” HTML íŒŒì¼ ìˆ˜ì§‘ ì¤‘...\n")
    html_files = []
    
    for year_folder in sorted(approval_folder.iterdir()):
        if year_folder.is_dir():
            year = year_folder.name
            year_html_files = list(year_folder.glob("*.html"))
            html_files.extend(year_html_files)
            print(f"   ğŸ“ {year}ë…„: {len(year_html_files)}ê°œ íŒŒì¼")
    
    print(f"\nâœ… ì´ {len(html_files)}ê°œ HTML íŒŒì¼ ë°œê²¬\n")

    # html_files = html_files[:2]  # í…ŒìŠ¤íŠ¸  #ì—¬ê¸°ë¥¼ ìˆ˜ì •í•˜ì„¸ìš”
    
    # 4. HTML íŒŒì¼ ì²˜ë¦¬
    print(f"ğŸ”„ HTML íŒŒì¼ ì²˜ë¦¬ ì‹œì‘...\n")
    result_dict = {}
    success_count = 0
    fail_count = 0
    total_original_size = 0
    total_minified_size = 0
    
    for i, html_file in enumerate(html_files, 1):
        try:
            # ë¬¸ì„œID ì¶”ì¶œ
            doc_id = extract_doc_id(html_file.name)
            if not doc_id:
                print(f"âš ï¸  [{i}/{len(html_files)}] ë¬¸ì„œID ì¶”ì¶œ ì‹¤íŒ¨: {html_file.name}")
                fail_count += 1
                continue
            
            # HTML íŒŒì¼ ì½ê¸°
            with open(html_file, 'r', encoding='utf-8', errors='ignore') as f:
                html_content = f.read()
            
            # 1. CSS inline ì‚½ì…
            html_with_css = insert_css_inline(html_content, css_dict)
            
            # 2. íŠ¹ì • HTML ì„¹ì…˜ ì‚­ì œ
            html_cleaned = remove_specific_sections(html_with_css)
            
            # 3. ì¶”ê°€ CSSë¥¼ ê° <style> íƒœê·¸ ë§¨ ì•ì— ì‚½ì…
            html_with_custom_css = add_custom_css_to_styles(html_cleaned)
            
            # 4. ëª¨ë“  <style> íƒœê·¸ë¥¼ </body> ì•ìœ¼ë¡œ ì´ë™ (body íƒœê·¸ê°€ ì•„ì§ ìˆìŒ)
            html_with_moved_styles = move_styles_before_body_close(html_with_custom_css)
            
            # 5. HTML wrapper íƒœê·¸ ì œê±° (ì´ì œ body íƒœê·¸ ì‚­ì œ)
            html_without_wrapper = remove_html_wrapper_tags(html_with_moved_styles)
            
            # 6. HTML ì••ì¶•
            minified_html = minify_html(html_without_wrapper)
            
            # ë”•ì…”ë„ˆë¦¬ì— ì¶”ê°€
            result_dict[doc_id] = minified_html
            
            success_count += 1
            
            # ì••ì¶• ë¹„ìœ¨ ê³„ì‚°
            original_size = len(html_without_wrapper)
            minified_size = len(minified_html)
            total_original_size += original_size
            total_minified_size += minified_size
            
            if i % 100 == 0 or i == len(html_files):
                print(f"âœ“ [{i}/{len(html_files)}] ì§„í–‰ ì¤‘... (ì„±ê³µ: {success_count}, ì‹¤íŒ¨: {fail_count})")
        
        except Exception as e:
            print(f"âœ— [{i}/{len(html_files)}] ì—ëŸ¬ ë°œìƒ: {html_file.name}")
            print(f"   ì—ëŸ¬ ë‚´ìš©: {str(e)}")
            fail_count += 1
            continue
    
    # 5. DB ì—…ë°ì´íŠ¸
    print(f"\n" + "="*60)
    update_db_docbody(result_dict, db_config, end_year)
    print(f"="*60)
    
    # 6. ê²°ê³¼ ìš”ì•½
    print(f"\n" + "="*60)
    print(f"âœ… ì™„ë£Œ!")
    print(f"="*60)
    print(f"ğŸ“Š í†µê³„:")
    print(f"   - ì „ì²´ HTML íŒŒì¼: {len(html_files)}ê°œ")
    print(f"   - ì„±ê³µ: {success_count}ê°œ")
    print(f"   - ì‹¤íŒ¨: {fail_count}ê°œ")
    print(f"   - ì´ ì›ë³¸ í¬ê¸°: {total_original_size:,} bytes")
    print(f"   - ì´ ì••ì¶• í¬ê¸°: {total_minified_size:,} bytes")
    if total_original_size > 0:
        total_ratio = (1 - total_minified_size/total_original_size) * 100
        print(f"   - ì „ì²´ ì••ì¶•ë¥ : â–¼{total_ratio:.1f}%")

if __name__ == "__main__":
    main()
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
HTML íŒŒì¼ì—ì„œ CSSë¥¼ inlineìœ¼ë¡œ ì‚½ì…í•˜ê³  ì••ì¶•í•˜ì—¬ DBì— ì €ì¥
"""
import json
import re
from pathlib import Path
from bs4 import BeautifulSoup
import pymysql

def minify_css(css):
    """CSSë¥¼ í•œ ì¤„ë¡œ ì••ì¶• (ë Œë”ë§ ê²°ê³¼ëŠ” ë™ì¼)"""
    # CSS ì£¼ì„ ì œê±°
    css = re.sub(r'/\*.*?\*/', '', css, flags=re.DOTALL)
    # ì¤„ë°”ê¿ˆ ì œê±°
    css = css.replace('\n', '').replace('\r', '')
    # ì—°ì†ëœ ê³µë°±ì„ í•˜ë‚˜ë¡œ
    css = re.sub(r'\s+', ' ', css)
    # { } : ; ì•ë’¤ ê³µë°± ì œê±°
    css = re.sub(r'\s*{\s*', '{', css)
    css = re.sub(r'\s*}\s*', '}', css)
    css = re.sub(r'\s*:\s*', ':', css)
    css = re.sub(r'\s*;\s*', ';', css)
    css = re.sub(r'\s*,\s*', ',', css)
    return css.strip()

def minify_html(html):
    """
    HTMLì„ ì™„ì „íˆ í•œ ì¤„ë¡œ ì••ì¶• (CSS í¬í•¨)
    - <pre>, <textarea> íƒœê·¸ ë‚´ë¶€ë§Œ ë³´ì¡´
    - CSS, Scriptë„ ëª¨ë‘ í•œ ì¤„ë¡œ ì••ì¶•
    """
    # 1. <style> íƒœê·¸ ë‚´ë¶€ CSSë¥¼ ì••ì¶•
    def minify_style_tag(match):
        tag_open = match.group(1)  # <style...>
        css_content = match.group(2)
        minified_css = minify_css(css_content)
        return f"{tag_open}{minified_css}</style>"
    
    html = re.sub(r'(<style[^>]*>)(.*?)</style>', minify_style_tag, html, flags=re.DOTALL | re.IGNORECASE)
    
    # 2. <script> íƒœê·¸ ë‚´ë¶€ë¥¼ í•œ ì¤„ë¡œ ì••ì¶•
    def minify_script_tag(match):
        tag_open = match.group(1)
        script_content = match.group(2)
        # ìŠ¤í¬ë¦½íŠ¸ëŠ” ê°„ë‹¨íˆ ì¤„ë°”ê¿ˆê³¼ ì—°ì† ê³µë°±ë§Œ ì œê±°
        minified = script_content.replace('\n', '').replace('\r', '')
        minified = re.sub(r'\s+', ' ', minified)
        return f"{tag_open}{minified}</script>"
    
    html = re.sub(r'(<script[^>]*>)(.*?)</script>', minify_script_tag, html, flags=re.DOTALL | re.IGNORECASE)
    
    # 3. <pre>, <textarea> íƒœê·¸ ë‚´ë¶€ë¥¼ ì„ì‹œë¡œ ë³´í˜¸
    pre_blocks = []
    def save_pre(match):
        pre_blocks.append(match.group(0))
        return f"___PRE_BLOCK_{len(pre_blocks)-1}___"
    
    html = re.sub(r'<pre[^>]*>.*?</pre>', save_pre, html, flags=re.DOTALL | re.IGNORECASE)
    html = re.sub(r'<textarea[^>]*>.*?</textarea>', save_pre, html, flags=re.DOTALL | re.IGNORECASE)
    
    # 4. HTML ì£¼ì„ ì œê±°
    html = re.sub(r'<!--.*?-->', '', html, flags=re.DOTALL)
    
    # 5. íƒœê·¸ ì‚¬ì´ì˜ ë¶ˆí•„ìš”í•œ ê³µë°± ì œê±°
    html = re.sub(r'>\s+<', '><', html)
    
    # 6. ì—°ì†ëœ ê³µë°±ì„ í•˜ë‚˜ë¡œ
    html = re.sub(r'\s+', ' ', html)
    
    # 7. ì¤„ë°”ê¿ˆ ì œê±°
    html = html.replace('\n', '').replace('\r', '')
    
    # 8. ë³´í˜¸í–ˆë˜ pre/textarea ë¸”ë¡ë“¤ ë³µì›
    for i, block in enumerate(pre_blocks):
        html = html.replace(f"___PRE_BLOCK_{i}___", block)
    
    return html.strip()

def extract_doc_id(filename):
    """íŒŒì¼ëª…ì—ì„œ ë¬¸ì„œID ì¶”ì¶œ (ì˜ˆ: 20151224_ì œëª©_2009465.html -> 2009465)"""
    # .html ì œê±°
    name_without_ext = filename.replace('.html', '')
    # ë§ˆì§€ë§‰ ì–¸ë”ìŠ¤ì½”ì–´ ë’¤ì˜ ìˆ«ì ë¶€ë¶„ì´ ë¬¸ì„œID
    parts = name_without_ext.split('_')
    if len(parts) >= 2:
        # ë§ˆì§€ë§‰ ë¶€ë¶„ì´ ìˆ«ìì¸ì§€ í™•ì¸
        last_part = parts[-1]
        if last_part.isdigit():
            return last_part
    return None

def read_css_files(css_folder):
    """CSS íŒŒì¼ 3ê°œë¥¼ ì½ì–´ì„œ ë”•ì…”ë„ˆë¦¬ë¡œ ë°˜í™˜"""
    css_dict = {}
    css_files = [
        'apr.doc.print.jstl.css',
        'font-load.css',
        'jquery.qtip.min.css'
    ]
    
    for css_file in css_files:
        css_path = css_folder / css_file
        if css_path.exists():
            with open(css_path, 'r', encoding='utf-8', errors='ignore') as f:
                css_dict[css_file] = f.read()
            print(f"âœ“ CSS íŒŒì¼ ì½ê¸° ì™„ë£Œ: {css_file}")
        else:
            print(f"âš ï¸  CSS íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {css_file}")
    
    return css_dict

def insert_css_inline(html_content, css_dict):
    """HTMLì˜ <link> íƒœê·¸ë¥¼ <style> íƒœê·¸ë¡œ êµì²´"""
    soup = BeautifulSoup(html_content, 'html.parser')
    
    # ëª¨ë“  <link> íƒœê·¸ ì°¾ê¸°
    link_tags = soup.find_all('link', rel='stylesheet')
    
    for link_tag in link_tags:
        href = link_tag.get('href', '')
        
        # CSS íŒŒì¼ëª… ì¶”ì¶œ
        css_filename = None
        if 'apr.doc.print.jstl.css' in href:
            css_filename = 'apr.doc.print.jstl.css'
        elif 'font-load.css' in href:
            css_filename = 'font-load.css'
        elif 'jquery.qtip.min.css' in href:
            css_filename = 'jquery.qtip.min.css'
        
        # í•´ë‹¹ CSSê°€ ìˆìœ¼ë©´ <style> íƒœê·¸ë¡œ êµì²´
        if css_filename and css_filename in css_dict:
            style_tag = soup.new_tag('style')
            style_tag.string = css_dict[css_filename]
            link_tag.replace_with(style_tag)
    
    return str(soup)

def remove_specific_sections(html_content):
    """ì²¨ë¶€íŒŒì¼, ì°¸ì¡°ë¬¸ì„œ, ê²°ì¬ì˜ê²¬, ê²°ì¬ëŒ“ê¸€ ì„¹ì…˜ ì‚­ì œ"""
    soup = BeautifulSoup(html_content, 'html.parser')
    
    # 1. ì²¨ë¶€íŒŒì¼ ì˜ì—­ ì‚­ì œ (id="attachFileArea"ì¸ div)
    for div in soup.find_all('div', id='attachFileArea'):
        div.decompose()
    
    # 2. ì²¨ë¶€íŒŒì¼ ì˜ì—­ ì‚­ì œ (captionì´ "íŒŒì¼,ë¬¸ì„œ"ì¸ ê²½ìš°)
    for div in soup.find_all('div', class_='table05_box'):
        caption = div.find('caption')
        if caption and 'íŒŒì¼' in caption.get_text(strip=True):
            div.decompose()
    
    # 3. ì²¨ë¶€íŒŒì¼ í–‰ ì‚­ì œ (thì— "ì²¨ë¶€íŒŒì¼" í…ìŠ¤íŠ¸ê°€ ìˆëŠ” tr)
    for th in soup.find_all('th'):
        if 'ì²¨ë¶€íŒŒì¼' in th.get_text(strip=True):
            tr = th.find_parent('tr')
            if tr:
                tr.decompose()
    
    # 4. ì°¸ì¡°ë¬¸ì„œ í–‰ ì‚­ì œ (thì— "ì°¸ì¡°ë¬¸ì„œ" í…ìŠ¤íŠ¸ê°€ ìˆëŠ” tr)
    for th in soup.find_all('th', scope='row'):
        if th.get_text(strip=True) == 'ì°¸ì¡°ë¬¸ì„œ':
            tr = th.find_parent('tr')
            if tr:
                tr.decompose()
    
    # 5. ê²°ì¬ì˜ê²¬ ì˜ì—­ ì‚­ì œ
    for div in soup.find_all('div', class_='table04_box'):
        caption = div.find('caption')
        if caption and 'ê²°ì¬ì˜ê²¬' in caption.get_text(strip=True):
            div.decompose()
    
    # 6. ê²°ì¬ëŒ“ê¸€ ì˜ì—­ ì‚­ì œ
    for div in soup.find_all('div', class_='table04_box'):
        caption = div.find('caption')
        if caption and 'ê²°ì¬ëŒ“ê¸€' in caption.get_text(strip=True):
            div.decompose()
    
    return str(soup)

def add_custom_css_to_styles(html_content):
    """ëª¨ë“  <style> íƒœê·¸ ë§¨ ì•ì— ì»¤ìŠ¤í…€ CSS ì¶”ê°€"""
    soup = BeautifulSoup(html_content, 'html.parser')
    
    custom_css = '.content{width: 80% !important;} .content table {width: 100% !important;} .content_box{width: 97.7% !important;}'
    
    # ëª¨ë“  <style> íƒœê·¸ ì°¾ê¸°
    style_tags = soup.find_all('style')
    
    for style_tag in style_tags:
        # ê¸°ì¡´ ë‚´ìš© ì•ì— ì»¤ìŠ¤í…€ CSS ì¶”ê°€
        original_content = style_tag.string if style_tag.string else ''
        style_tag.string = custom_css + original_content
    
    return str(soup)

def move_styles_before_body_close(html_content):
    """ëª¨ë“  <style> íƒœê·¸ë¥¼ </body> íƒœê·¸ ì§ì „ìœ¼ë¡œ ì´ë™"""
    soup = BeautifulSoup(html_content, 'html.parser')
    
    # ëª¨ë“  <style> íƒœê·¸ ì°¾ê¸°
    style_tags = soup.find_all('style')
    
    # body íƒœê·¸ ì°¾ê¸°
    body = soup.find('body')
    
    if body and style_tags:
        # ê¸°ì¡´ ìœ„ì¹˜ì—ì„œ style íƒœê·¸ë“¤ì„ ì œê±°í•˜ê³  ìˆ˜ì§‘
        for style_tag in style_tags:
            style_tag.extract()
        
        # body íƒœê·¸ ëì— ì¶”ê°€
        for style_tag in style_tags:
            body.append(style_tag)
    
    return str(soup)

def check_and_fix_column_type(db_config):
    """doc_body ì»¬ëŸ¼ íƒ€ì… í™•ì¸ ë° í•„ìš”ì‹œ ë³€ê²½"""
    conn = None
    cursor = None
    
    try:
        conn = pymysql.connect(
            host=db_config['host'],
            user=db_config['user'],
            password=db_config['password'],
            database=db_config['database'],
            charset='utf8mb4'
        )
        cursor = conn.cursor()
        
        # ì»¬ëŸ¼ íƒ€ì… í™•ì¸
        cursor.execute("SHOW COLUMNS FROM documents LIKE 'doc_body'")
        result = cursor.fetchone()
        
        if result:
            column_type = result[1].decode('utf-8') if isinstance(result[1], bytes) else result[1]
            print(f"\n=== doc_body ì»¬ëŸ¼ íƒ€ì… í™•ì¸ ===")
            print(f"í˜„ì¬ íƒ€ì…: {column_type}")
            
            # TEXT íƒ€ì…ì´ë©´ MEDIUMTEXTë¡œ ë³€ê²½
            if column_type.upper() in ['TEXT', 'VARCHAR']:
                print(f"âš ï¸  {column_type}ëŠ” í¬ê¸°ê°€ ë¶€ì¡±í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.")
                print(f"ğŸ”„ MEDIUMTEXTë¡œ ë³€ê²½ ì¤‘...")
                
                cursor.execute("ALTER TABLE documents MODIFY COLUMN doc_body MEDIUMTEXT")
                conn.commit()
                print(f"âœ… doc_body ì»¬ëŸ¼ì„ MEDIUMTEXTë¡œ ë³€ê²½ ì™„ë£Œ!")
            elif column_type.upper() in ['MEDIUMTEXT', 'LONGTEXT']:
                print(f"âœ… ì»¬ëŸ¼ íƒ€ì…ì´ ì¶©ë¶„í•©ë‹ˆë‹¤.")
            else:
                print(f"âš ï¸  ì˜ˆìƒì¹˜ ëª»í•œ íƒ€ì…ì…ë‹ˆë‹¤: {column_type}")
        
    except Exception as e:
        print(f"âŒ ì»¬ëŸ¼ íƒ€ì… í™•ì¸/ë³€ê²½ ì¤‘ ì˜¤ë¥˜: {e}")
    
    finally:
        if cursor:
            cursor.close()
        if conn:
            conn.close()

def update_db_docbody(processed_data, db_config, end_year):
    """ì²˜ë¦¬ëœ HTMLì„ DBì˜ doc_body ì»¬ëŸ¼ì— ì—…ë°ì´íŠ¸"""
    conn = None
    cursor = None
    
    try:
        print("\n=== DB ì—°ê²° ì‹œì‘ ===")
        print(f"Host: {db_config['host']}, Database: {db_config['database']}")
        
        # DB ì—°ê²°
        conn = pymysql.connect(
            host=db_config['host'],
            user=db_config['user'],
            password=db_config['password'],
            database=db_config['database'],
            charset='utf8mb4'
        )
        print("âœ“ DB ì—°ê²° ì„±ê³µ")
        
        cursor = conn.cursor()
        
        # ì—…ë°ì´íŠ¸ ì¿¼ë¦¬
        update_sql = """
        UPDATE documents 
        SET doc_body = %s 
        WHERE source_id = %s AND end_year = %s
        """
        
        success_count = 0
        fail_count = 0
        failed_items = []  # ì‹¤íŒ¨í•œ í•­ëª© ì €ì¥
        
        print(f"\n=== DB ì—…ë°ì´íŠ¸ ì‹œì‘ ({len(processed_data)}ê±´) ===")
        
        for idx, (doc_id, html_content) in enumerate(processed_data.items(), 1):
            try:
                html_size = len(html_content.encode('utf-8'))
                cursor.execute(update_sql, (html_content, doc_id, end_year))
                success_count += 1
                
                # 100ê±´ë§ˆë‹¤ ì¶œë ¥
                if idx % 100 == 0 or idx == len(processed_data):
                    print(f"  ì§„í–‰: {idx}/{len(processed_data)} (ì„±ê³µ: {success_count}, ì‹¤íŒ¨: {fail_count})")
                    
            except Exception as e:
                fail_count += 1
                html_size = len(html_content.encode('utf-8'))
                failed_items.append({
                    'doc_id': doc_id,
                    'size': html_size,
                    'error': str(e)
                })
                
                if fail_count <= 10:
                    print(f"  [ì˜¤ë¥˜ {fail_count}] source_id: {doc_id}, í¬ê¸°: {html_size:,} bytes - {str(e)[:80]}")
        
        # ì»¤ë°‹
        print("\n=== ì»¤ë°‹ ì‹œì‘ ===")
        conn.commit()
        print(f"âœ“ ì»¤ë°‹ ì™„ë£Œ")
        
        # ê²°ê³¼ ìš”ì•½
        print(f"\n=== DB ì—…ë°ì´íŠ¸ ê²°ê³¼ ===")
        print(f"âœ“ ì„±ê³µ: {success_count}ê±´")
        print(f"âœ— ì‹¤íŒ¨: {fail_count}ê±´")
        
        # ì‹¤íŒ¨í•œ í•­ëª© ìƒì„¸ ì •ë³´
        if failed_items:
            print(f"\n=== ì‹¤íŒ¨ í•­ëª© ë¶„ì„ ===")
            sizes = [item['size'] for item in failed_items]
            print(f"ì‹¤íŒ¨í•œ í•­ëª© í‰ê·  í¬ê¸°: {sum(sizes)/len(sizes):,.0f} bytes")
            print(f"ì‹¤íŒ¨í•œ í•­ëª© ìµœëŒ€ í¬ê¸°: {max(sizes):,} bytes")
            print(f"ì‹¤íŒ¨í•œ í•­ëª© ìµœì†Œ í¬ê¸°: {min(sizes):,} bytes")
            
            # ì‹¤íŒ¨ í•­ëª©ì„ íŒŒì¼ë¡œ ì €ì¥
            with open('failed_updates.txt', 'w', encoding='utf-8') as f:
                for item in failed_items:
                    f.write(f"source_id: {item['doc_id']}, size: {item['size']:,} bytes, error: {item['error']}\n")
            print(f"âœ“ ì‹¤íŒ¨ í•­ëª© ìƒì„¸ ì •ë³´ë¥¼ 'failed_updates.txt'ì— ì €ì¥í–ˆìŠµë‹ˆë‹¤.")
        
    except pymysql.Error as e:
        print(f"\nâŒ DB ì˜¤ë¥˜ ë°œìƒ:")
        print(f"  Error Code: {e.args[0]}")
        print(f"  Error Message: {e.args[1]}")
        
    except Exception as e:
        print(f"\nâŒ ì¼ë°˜ ì˜¤ë¥˜ ë°œìƒ: {e}")
        
    finally:
        if cursor:
            cursor.close()
        if conn:
            conn.close()
            print("\nâœ“ DB ì—°ê²° ì¢…ë£Œ")

def main():
    # 1. ê²½ë¡œ ì„¤ì • (ì‚¬ìš©ì í™˜ê²½ì— ë§ê²Œ ìˆ˜ì • í•„ìš”)
    
    base_path = Path("C:/Users/LEEJUHWAN/Downloads/2021-01-01~2025-10-31/html") #ì—¬ê¸°ë¥¼ ìˆ˜ì •í•˜ì„¸ìš”
    end_year = 2025  #ì—¬ê¸°ë¥¼ ìˆ˜ì •í•˜ì„¸ìš” - í´ë” ê¸°ê°„ì— ë§ê²Œ ìˆ˜ì •
    approval_folder = base_path / "ê²°ì¬"
    css_folder = base_path / "resource" / "css"
    
    print(f"ğŸ“‚ ê¸°ë³¸ ê²½ë¡œ: {base_path}")
    print(f"ğŸ“‚ ê²°ì¬ í´ë”: {approval_folder}")
    print(f"ğŸ“‚ CSS í´ë”: {css_folder}\n")
    
    # ê²½ë¡œ ì¡´ì¬ í™•ì¸
    if not approval_folder.exists():
        print(f"âŒ ê²°ì¬ í´ë”ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {approval_folder}")
        print(f"ğŸ’¡ base_path ë³€ìˆ˜ë¥¼ ì‹¤ì œ ê²½ë¡œë¡œ ìˆ˜ì •í•´ì£¼ì„¸ìš”!")
        return
    
    if not css_folder.exists():
        print(f"âŒ CSS í´ë”ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {css_folder}")
        return
    
    # DB ì„¤ì •
    db_config = {
        'host': 'localhost',
        'user': 'root',
        'password': '1234',
        'database': 'any_approval' #ì—¬ê¸°ë¥¼ ìˆ˜ì •í•˜ì„¸ìš” - í†µí•© DBëª…
    }
    
    # DB ì»¬ëŸ¼ íƒ€ì… í™•ì¸ ë° í•„ìš”ì‹œ ë³€ê²½
    check_and_fix_column_type(db_config)
    
    # 2. CSS íŒŒì¼ ì½ê¸°
    print(f"\nğŸ”„ CSS íŒŒì¼ ì½ëŠ” ì¤‘...\n")
    css_dict = read_css_files(css_folder)
    
    if not css_dict:
        print(f"âŒ CSS íŒŒì¼ì„ ì½ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        return
    
    print(f"\nâœ… {len(css_dict)}ê°œ CSS íŒŒì¼ ë¡œë“œ ì™„ë£Œ\n")
    
    # 3. ì—°ë„ë³„ í´ë”ì—ì„œ HTML íŒŒì¼ ìˆ˜ì§‘
    print(f"ğŸ” HTML íŒŒì¼ ìˆ˜ì§‘ ì¤‘...\n")
    html_files = []
    
    for year_folder in sorted(approval_folder.iterdir()):
        if year_folder.is_dir():
            year = year_folder.name
            year_html_files = list(year_folder.glob("*.html"))
            html_files.extend(year_html_files)
            print(f"   ğŸ“ {year}ë…„: {len(year_html_files)}ê°œ íŒŒì¼")
    
    print(f"\nâœ… ì´ {len(html_files)}ê°œ HTML íŒŒì¼ ë°œê²¬\n")

    # html_files = html_files[:2]  # í…ŒìŠ¤íŠ¸  #ì—¬ê¸°ë¥¼ ìˆ˜ì •í•˜ì„¸ìš”
    
    # 4. HTML íŒŒì¼ ì²˜ë¦¬
    print(f"ğŸ”„ HTML íŒŒì¼ ì²˜ë¦¬ ì‹œì‘...\n")
    result_dict = {}
    success_count = 0
    fail_count = 0
    total_original_size = 0
    total_minified_size = 0
    
    for i, html_file in enumerate(html_files, 1):
        try:
            # ë¬¸ì„œID ì¶”ì¶œ
            doc_id = extract_doc_id(html_file.name)
            if not doc_id:
                print(f"âš ï¸  [{i}/{len(html_files)}] ë¬¸ì„œID ì¶”ì¶œ ì‹¤íŒ¨: {html_file.name}")
                fail_count += 1
                continue
            
            # HTML íŒŒì¼ ì½ê¸°
            with open(html_file, 'r', encoding='utf-8', errors='ignore') as f:
                html_content = f.read()
            
            # 1. CSS inline ì‚½ì…
            html_with_css = insert_css_inline(html_content, css_dict)
            
            # 2. íŠ¹ì • HTML ì„¹ì…˜ ì‚­ì œ
            html_cleaned = remove_specific_sections(html_with_css)
            
            # 3. ì¶”ê°€ CSSë¥¼ ê° <style> íƒœê·¸ ë§¨ ì•ì— ì‚½ì…
            html_with_custom_css = add_custom_css_to_styles(html_cleaned)
            
            # 4. ëª¨ë“  <style> íƒœê·¸ë¥¼ </body> ì•ìœ¼ë¡œ ì´ë™
            html_with_moved_styles = move_styles_before_body_close(html_with_custom_css)
            
            # 5. HTML ì••ì¶•
            minified_html = minify_html(html_with_moved_styles)
            
            # ë”•ì…”ë„ˆë¦¬ì— ì¶”ê°€
            result_dict[doc_id] = minified_html
            
            success_count += 1
            
            # ì••ì¶• ë¹„ìœ¨ ê³„ì‚°
            original_size = len(html_with_moved_styles)
            minified_size = len(minified_html)
            total_original_size += original_size
            total_minified_size += minified_size
            
            if i % 100 == 0 or i == len(html_files):
                print(f"âœ“ [{i}/{len(html_files)}] ì§„í–‰ ì¤‘... (ì„±ê³µ: {success_count}, ì‹¤íŒ¨: {fail_count})")
        
        except Exception as e:
            print(f"âœ— [{i}/{len(html_files)}] ì—ëŸ¬ ë°œìƒ: {html_file.name}")
            print(f"   ì—ëŸ¬ ë‚´ìš©: {str(e)}")
            fail_count += 1
            continue
    
    # 5. DB ì—…ë°ì´íŠ¸
    print(f"\n" + "="*60)
    update_db_docbody(result_dict, db_config, end_year)
    print(f"="*60)
    
    # 6. ê²°ê³¼ ìš”ì•½
    print(f"\n" + "="*60)
    print(f"âœ… ì™„ë£Œ!")
    print(f"="*60)
    print(f"ğŸ“Š í†µê³„:")
    print(f"   - ì „ì²´ HTML íŒŒì¼: {len(html_files)}ê°œ")
    print(f"   - ì„±ê³µ: {success_count}ê°œ")
    print(f"   - ì‹¤íŒ¨: {fail_count}ê°œ")
    print(f"   - ì´ ì›ë³¸ í¬ê¸°: {total_original_size:,} bytes")
    print(f"   - ì´ ì••ì¶• í¬ê¸°: {total_minified_size:,} bytes")
    if total_original_size > 0:
        total_ratio = (1 - total_minified_size/total_original_size) * 100
        print(f"   - ì „ì²´ ì••ì¶•ë¥ : â–¼{total_ratio:.1f}%")

if __name__ == "__main__":
    main()
```

```python
#ì¶”ê°€
# 1. CSS inline ì‚½ì… (ê¸°ì¡´ ì½”ë“œ)
# 2. íŠ¹ì • HTML ì„¹ì…˜ ì‚­ì œ (ìƒˆ ê¸°ëŠ¥)
# 3. ì¶”ê°€ CSSë¥¼ ê° <style> íƒœê·¸ ë§¨ ì•ì— ì‚½ì… (ìƒˆ ê¸°ëŠ¥)
# 4. ëª¨ë“  <style> íƒœê·¸ë¥¼ </body> ì•ìœ¼ë¡œ ì´ë™ (ìƒˆ ê¸°ëŠ¥)
# 5. HTML ì••ì¶• (ê¸°ì¡´ ì½”ë“œ)

#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
HTML íŒŒì¼ì—ì„œ CSSë¥¼ inlineìœ¼ë¡œ ì‚½ì…í•˜ê³  ì••ì¶•í•˜ì—¬ ë”•ì…”ë„ˆë¦¬ë¡œ ë³€í™˜
"""
import json
import re
from pathlib import Path
from bs4 import BeautifulSoup

def minify_css(css):
    """CSSë¥¼ í•œ ì¤„ë¡œ ì••ì¶• (ë Œë”ë§ ê²°ê³¼ëŠ” ë™ì¼)"""
    # CSS ì£¼ì„ ì œê±°
    css = re.sub(r'/\*.*?\*/', '', css, flags=re.DOTALL)
    # ì¤„ë°”ê¿ˆ ì œê±°
    css = css.replace('\n', '').replace('\r', '')
    # ì—°ì†ëœ ê³µë°±ì„ í•˜ë‚˜ë¡œ
    css = re.sub(r'\s+', ' ', css)
    # { } : ; ì•ë’¤ ê³µë°± ì œê±°
    css = re.sub(r'\s*{\s*', '{', css)
    css = re.sub(r'\s*}\s*', '}', css)
    css = re.sub(r'\s*:\s*', ':', css)
    css = re.sub(r'\s*;\s*', ';', css)
    css = re.sub(r'\s*,\s*', ',', css)
    return css.strip()

def minify_html(html):
    """
    HTMLì„ ì™„ì „íˆ í•œ ì¤„ë¡œ ì••ì¶• (CSS í¬í•¨)
    - <pre>, <textarea> íƒœê·¸ ë‚´ë¶€ë§Œ ë³´ì¡´
    - CSS, Scriptë„ ëª¨ë‘ í•œ ì¤„ë¡œ ì••ì¶•
    """
    # 1. <style> íƒœê·¸ ë‚´ë¶€ CSSë¥¼ ì••ì¶•
    def minify_style_tag(match):
        tag_open = match.group(1)  # <style...>
        css_content = match.group(2)
        minified_css = minify_css(css_content)
        return f"{tag_open}{minified_css}</style>"
    
    html = re.sub(r'(<style[^>]*>)(.*?)</style>', minify_style_tag, html, flags=re.DOTALL | re.IGNORECASE)
    
    # 2. <script> íƒœê·¸ ë‚´ë¶€ë¥¼ í•œ ì¤„ë¡œ ì••ì¶•
    def minify_script_tag(match):
        tag_open = match.group(1)
        script_content = match.group(2)
        # ìŠ¤í¬ë¦½íŠ¸ëŠ” ê°„ë‹¨íˆ ì¤„ë°”ê¿ˆê³¼ ì—°ì† ê³µë°±ë§Œ ì œê±°
        minified = script_content.replace('\n', '').replace('\r', '')
        minified = re.sub(r'\s+', ' ', minified)
        return f"{tag_open}{minified}</script>"
    
    html = re.sub(r'(<script[^>]*>)(.*?)</script>', minify_script_tag, html, flags=re.DOTALL | re.IGNORECASE)
    
    # 3. <pre>, <textarea> íƒœê·¸ ë‚´ë¶€ë¥¼ ì„ì‹œë¡œ ë³´í˜¸
    pre_blocks = []
    def save_pre(match):
        pre_blocks.append(match.group(0))
        return f"___PRE_BLOCK_{len(pre_blocks)-1}___"
    
    html = re.sub(r'<pre[^>]*>.*?</pre>', save_pre, html, flags=re.DOTALL | re.IGNORECASE)
    html = re.sub(r'<textarea[^>]*>.*?</textarea>', save_pre, html, flags=re.DOTALL | re.IGNORECASE)
    
    # 4. HTML ì£¼ì„ ì œê±°
    html = re.sub(r'<!--.*?-->', '', html, flags=re.DOTALL)
    
    # 5. íƒœê·¸ ì‚¬ì´ì˜ ë¶ˆí•„ìš”í•œ ê³µë°± ì œê±°
    html = re.sub(r'>\s+<', '><', html)
    
    # 6. ì—°ì†ëœ ê³µë°±ì„ í•˜ë‚˜ë¡œ
    html = re.sub(r'\s+', ' ', html)
    
    # 7. ì¤„ë°”ê¿ˆ ì œê±°
    html = html.replace('\n', '').replace('\r', '')
    
    # 8. ë³´í˜¸í–ˆë˜ pre/textarea ë¸”ë¡ë“¤ ë³µì›
    for i, block in enumerate(pre_blocks):
        html = html.replace(f"___PRE_BLOCK_{i}___", block)
    
    return html.strip()

def extract_doc_id(filename):
    """íŒŒì¼ëª…ì—ì„œ ë¬¸ì„œID ì¶”ì¶œ (ì˜ˆ: 20151224_ì œëª©_2009465.html -> 2009465)"""
    # .html ì œê±°
    name_without_ext = filename.replace('.html', '')
    # ë§ˆì§€ë§‰ ì–¸ë”ìŠ¤ì½”ì–´ ë’¤ì˜ ìˆ«ì ë¶€ë¶„ì´ ë¬¸ì„œID
    parts = name_without_ext.split('_')
    if len(parts) >= 2:
        # ë§ˆì§€ë§‰ ë¶€ë¶„ì´ ìˆ«ìì¸ì§€ í™•ì¸
        last_part = parts[-1]
        if last_part.isdigit():
            return last_part
    return None

def read_css_files(css_folder):
    """CSS íŒŒì¼ 3ê°œë¥¼ ì½ì–´ì„œ ë”•ì…”ë„ˆë¦¬ë¡œ ë°˜í™˜"""
    css_dict = {}
    css_files = [
        'apr.doc.print.jstl.css',
        'font-load.css',
        'jquery.qtip.min.css'
    ]
    
    for css_file in css_files:
        css_path = css_folder / css_file
        if css_path.exists():
            with open(css_path, 'r', encoding='utf-8', errors='ignore') as f:
                css_dict[css_file] = f.read()
            print(f"âœ“ CSS íŒŒì¼ ì½ê¸° ì™„ë£Œ: {css_file}")
        else:
            print(f"âš ï¸  CSS íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {css_file}")
    
    return css_dict

def insert_css_inline(html_content, css_dict):
    """HTMLì˜ <link> íƒœê·¸ë¥¼ <style> íƒœê·¸ë¡œ êµì²´"""
    soup = BeautifulSoup(html_content, 'html.parser')
    
    # ëª¨ë“  <link> íƒœê·¸ ì°¾ê¸°
    link_tags = soup.find_all('link', rel='stylesheet')
    
    for link_tag in link_tags:
        href = link_tag.get('href', '')
        
        # CSS íŒŒì¼ëª… ì¶”ì¶œ
        css_filename = None
        if 'apr.doc.print.jstl.css' in href:
            css_filename = 'apr.doc.print.jstl.css'
        elif 'font-load.css' in href:
            css_filename = 'font-load.css'
        elif 'jquery.qtip.min.css' in href:
            css_filename = 'jquery.qtip.min.css'
        
        # í•´ë‹¹ CSSê°€ ìˆìœ¼ë©´ <style> íƒœê·¸ë¡œ êµì²´
        if css_filename and css_filename in css_dict:
            style_tag = soup.new_tag('style')
            style_tag.string = css_dict[css_filename]
            link_tag.replace_with(style_tag)
    
    return str(soup)

def remove_specific_sections(html_content):
    """ì²¨ë¶€íŒŒì¼, ì°¸ì¡°ë¬¸ì„œ, ê²°ì¬ì˜ê²¬, ê²°ì¬ëŒ“ê¸€ ì„¹ì…˜ ì‚­ì œ"""
    soup = BeautifulSoup(html_content, 'html.parser')
    
    # 1. ì²¨ë¶€íŒŒì¼ ì˜ì—­ ì‚­ì œ (id="attachFileArea"ì¸ div)
    for div in soup.find_all('div', id='attachFileArea'):
        div.decompose()
    
    # 2. ì²¨ë¶€íŒŒì¼ ì˜ì—­ ì‚­ì œ (captionì´ "íŒŒì¼,ë¬¸ì„œ"ì¸ ê²½ìš°)
    for div in soup.find_all('div', class_='table05_box'):
        caption = div.find('caption')
        if caption and 'íŒŒì¼' in caption.get_text(strip=True):
            div.decompose()
    
    # 3. ì²¨ë¶€íŒŒì¼ í–‰ ì‚­ì œ (thì— "ì²¨ë¶€íŒŒì¼" í…ìŠ¤íŠ¸ê°€ ìˆëŠ” tr)
    for th in soup.find_all('th'):
        if 'ì²¨ë¶€íŒŒì¼' in th.get_text(strip=True):
            tr = th.find_parent('tr')
            if tr:
                tr.decompose()
    
    # 4. ì°¸ì¡°ë¬¸ì„œ í–‰ ì‚­ì œ (thì— "ì°¸ì¡°ë¬¸ì„œ" í…ìŠ¤íŠ¸ê°€ ìˆëŠ” tr)
    for th in soup.find_all('th', scope='row'):
        if th.get_text(strip=True) == 'ì°¸ì¡°ë¬¸ì„œ':
            tr = th.find_parent('tr')
            if tr:
                tr.decompose()
    
    # 5. ê²°ì¬ì˜ê²¬ ì˜ì—­ ì‚­ì œ
    for div in soup.find_all('div', class_='table04_box'):
        caption = div.find('caption')
        if caption and 'ê²°ì¬ì˜ê²¬' in caption.get_text(strip=True):
            div.decompose()
    
    # 6. ê²°ì¬ëŒ“ê¸€ ì˜ì—­ ì‚­ì œ
    for div in soup.find_all('div', class_='table04_box'):
        caption = div.find('caption')
        if caption and 'ê²°ì¬ëŒ“ê¸€' in caption.get_text(strip=True):
            div.decompose()
    
    return str(soup)

def add_custom_css_to_styles(html_content):
    """ëª¨ë“  <style> íƒœê·¸ ë§¨ ì•ì— ì»¤ìŠ¤í…€ CSS ì¶”ê°€"""
    soup = BeautifulSoup(html_content, 'html.parser')
    
    custom_css = '.content{width: 80% !important;} .content table {width: 100% !important;} .content_box{width: 97.7% !important;}'
    
    # ëª¨ë“  <style> íƒœê·¸ ì°¾ê¸°
    style_tags = soup.find_all('style')
    
    for style_tag in style_tags:
        # ê¸°ì¡´ ë‚´ìš© ì•ì— ì»¤ìŠ¤í…€ CSS ì¶”ê°€
        original_content = style_tag.string if style_tag.string else ''
        style_tag.string = custom_css + original_content
    
    return str(soup)

def move_styles_before_body_close(html_content):
    """ëª¨ë“  <style> íƒœê·¸ë¥¼ </body> íƒœê·¸ ì§ì „ìœ¼ë¡œ ì´ë™"""
    soup = BeautifulSoup(html_content, 'html.parser')
    
    # ëª¨ë“  <style> íƒœê·¸ ì°¾ê¸°
    style_tags = soup.find_all('style')
    
    # body íƒœê·¸ ì°¾ê¸°
    body = soup.find('body')
    
    if body and style_tags:
        # ê¸°ì¡´ ìœ„ì¹˜ì—ì„œ style íƒœê·¸ë“¤ì„ ì œê±°í•˜ê³  ìˆ˜ì§‘
        for style_tag in style_tags:
            style_tag.extract()
        
        # body íƒœê·¸ ëì— ì¶”ê°€
        for style_tag in style_tags:
            body.append(style_tag)
    
    return str(soup)

def main():
    # 1. ê²½ë¡œ ì„¤ì • (ì‚¬ìš©ì í™˜ê²½ì— ë§ê²Œ ìˆ˜ì • í•„ìš”)
    
    base_path = Path("C:/Users/LEEJUHWAN/Downloads/2021-01-01~2025-10-31/html") #ì—¬ê¸°ë¥¼ ìˆ˜ì •í•˜ì„¸ìš”
    approval_folder = base_path / "ê²°ì¬"
    css_folder = base_path / "resource" / "css"
    
    print(f"ğŸ“‚ ê¸°ë³¸ ê²½ë¡œ: {base_path}")
    print(f"ğŸ“‚ ê²°ì¬ í´ë”: {approval_folder}")
    print(f"ğŸ“‚ CSS í´ë”: {css_folder}\n")
    
    # ê²½ë¡œ ì¡´ì¬ í™•ì¸
    if not approval_folder.exists():
        print(f"âŒ ê²°ì¬ í´ë”ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {approval_folder}")
        print(f"ğŸ’¡ base_path ë³€ìˆ˜ë¥¼ ì‹¤ì œ ê²½ë¡œë¡œ ìˆ˜ì •í•´ì£¼ì„¸ìš”!")
        return
    
    if not css_folder.exists():
        print(f"âŒ CSS í´ë”ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {css_folder}")
        return
    
    # 2. CSS íŒŒì¼ ì½ê¸°
    print(f"ğŸ”„ CSS íŒŒì¼ ì½ëŠ” ì¤‘...\n")
    css_dict = read_css_files(css_folder)
    
    if not css_dict:
        print(f"âŒ CSS íŒŒì¼ì„ ì½ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        return
    
    print(f"\nâœ… {len(css_dict)}ê°œ CSS íŒŒì¼ ë¡œë“œ ì™„ë£Œ\n")
    
    # 3. ì—°ë„ë³„ í´ë”ì—ì„œ HTML íŒŒì¼ ìˆ˜ì§‘
    print(f"ğŸ” HTML íŒŒì¼ ìˆ˜ì§‘ ì¤‘...\n")
    html_files = []
    
    for year_folder in sorted(approval_folder.iterdir()):
        if year_folder.is_dir():
            year = year_folder.name
            year_html_files = list(year_folder.glob("*.html"))
            html_files.extend(year_html_files)
            print(f"   ğŸ“ {year}ë…„: {len(year_html_files)}ê°œ íŒŒì¼")
    
    print(f"\nâœ… ì´ {len(html_files)}ê°œ HTML íŒŒì¼ ë°œê²¬\n")

    # html_files = html_files[:2]  # í…ŒìŠ¤íŠ¸  #ì—¬ê¸°ë¥¼ ìˆ˜ì •í•˜ì„¸ìš”
    
    # 4. HTML íŒŒì¼ ì²˜ë¦¬
    print(f"ğŸ”„ HTML íŒŒì¼ ì²˜ë¦¬ ì‹œì‘...\n")
    result_dict = {}
    success_count = 0
    fail_count = 0
    total_original_size = 0
    total_minified_size = 0
    
    for i, html_file in enumerate(html_files, 1):
        try:
            # ë¬¸ì„œID ì¶”ì¶œ
            doc_id = extract_doc_id(html_file.name)
            if not doc_id:
                print(f"âš ï¸  [{i}/{len(html_files)}] ë¬¸ì„œID ì¶”ì¶œ ì‹¤íŒ¨: {html_file.name}")
                fail_count += 1
                continue
            
            # HTML íŒŒì¼ ì½ê¸°
            with open(html_file, 'r', encoding='utf-8', errors='ignore') as f:
                html_content = f.read()
            
            # 1. CSS inline ì‚½ì…
            html_with_css = insert_css_inline(html_content, css_dict)
            
            # 2. íŠ¹ì • HTML ì„¹ì…˜ ì‚­ì œ
            html_cleaned = remove_specific_sections(html_with_css)
            
            # 3. ì¶”ê°€ CSSë¥¼ ê° <style> íƒœê·¸ ë§¨ ì•ì— ì‚½ì…
            html_with_custom_css = add_custom_css_to_styles(html_cleaned)
            
            # 4. ëª¨ë“  <style> íƒœê·¸ë¥¼ </body> ì•ìœ¼ë¡œ ì´ë™
            html_with_moved_styles = move_styles_before_body_close(html_with_custom_css)
            
            # 5. HTML ì••ì¶•
            minified_html = minify_html(html_with_moved_styles)
            
            # ë”•ì…”ë„ˆë¦¬ì— ì¶”ê°€
            result_dict[doc_id] = minified_html
            
            success_count += 1
            
            # ì••ì¶• ë¹„ìœ¨ ê³„ì‚°
            original_size = len(html_with_moved_styles)
            minified_size = len(minified_html)
            total_original_size += original_size
            total_minified_size += minified_size
            ratio = (1 - minified_size/original_size) * 100 if original_size > 0 else 0
            
            if i % 100 == 0 or i == len(html_files):
                print(f"âœ“ [{i}/{len(html_files)}] ì§„í–‰ ì¤‘... (ì„±ê³µ: {success_count}, ì‹¤íŒ¨: {fail_count})")
        
        except Exception as e:
            print(f"âœ— [{i}/{len(html_files)}] ì—ëŸ¬ ë°œìƒ: {html_file.name}")
            print(f"   ì—ëŸ¬ ë‚´ìš©: {str(e)}")
            fail_count += 1
            continue
    
    # 5. JSON íŒŒì¼ë¡œ ì €ì¥
    output_file = Path("two_htmltodocbodyconvert_result.json") #ì—¬ê¸°ë¥¼ ìˆ˜ì •í•˜ì„¸ìš”
    output_file.parent.mkdir(parents=True, exist_ok=True)
    
    print(f"\nğŸ’¾ JSON íŒŒì¼ ì €ì¥ ì¤‘...")
    with open(output_file, "w", encoding="utf-8") as f:
        json.dump(result_dict, f, ensure_ascii=False, indent=2)
    
    # 6. ê²°ê³¼ ìš”ì•½
    print(f"\n" + "="*60)
    print(f"âœ… ì™„ë£Œ!")
    print(f"="*60)
    print(f"ğŸ“Š í†µê³„:")
    print(f"   - ì „ì²´ HTML íŒŒì¼: {len(html_files)}ê°œ")
    print(f"   - ì„±ê³µ: {success_count}ê°œ")
    print(f"   - ì‹¤íŒ¨: {fail_count}ê°œ")
    print(f"   - ì´ ì›ë³¸ í¬ê¸°: {total_original_size:,} bytes")
    print(f"   - ì´ ì••ì¶• í¬ê¸°: {total_minified_size:,} bytes")
    if total_original_size > 0:
        total_ratio = (1 - total_minified_size/total_original_size) * 100
        print(f"   - ì „ì²´ ì••ì¶•ë¥ : â–¼{total_ratio:.1f}%")
    print(f"\nğŸ“ ì €ì¥ ìœ„ì¹˜: {output_file}")
    print(f"\nğŸ’¡ ì‚¬ìš© ì˜ˆì‹œ:")
    print(f"   import json")
    print(f"   with open('{output_file.name}', 'r', encoding='utf-8') as f:")
    print(f"       data = json.load(f)") 
    print(f"   html = data['2009465']  # ë¬¸ì„œIDë¡œ ë°”ë¡œ ì ‘ê·¼")

if __name__ == "__main__":
    main()
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
HTML íŒŒì¼ì—ì„œ CSSë¥¼ inlineìœ¼ë¡œ ì‚½ì…í•˜ê³  ì••ì¶•í•˜ì—¬ ë”•ì…”ë„ˆë¦¬ë¡œ ë³€í™˜

ìŠ¤íƒ€ì¼íƒœê·¸ë¥¼ </body>ì•ìœ¼ë¡œ ì˜®ê¸°ê¸° . 
ì²¨ë¶€íŒŒì¼~ê²°ì¬ëŒ“ê¸€ ì‚­ì œ(html)
ì•„ë˜ë¥¼ <style> ë°”ë¡œ ë’¤ì— ë¶™ì´ê¸°, 
.content{width: 80% !important;} .content table {width: 100% !important;} .content_box{width: 97.7% !important;}ì´ê±° ìŠ¤íƒ€ì¼íƒœê·¸ ì•ˆì—ì„œ ë§¨ì•ì— ì¶”ê°€
"""
import json
import re
from pathlib import Path
from bs4 import BeautifulSoup

def minify_css(css):
    """CSSë¥¼ í•œ ì¤„ë¡œ ì••ì¶• (ë Œë”ë§ ê²°ê³¼ëŠ” ë™ì¼)"""
    # CSS ì£¼ì„ ì œê±°
    css = re.sub(r'/\*.*?\*/', '', css, flags=re.DOTALL)
    # ì¤„ë°”ê¿ˆ ì œê±°
    css = css.replace('\n', '').replace('\r', '')
    # ì—°ì†ëœ ê³µë°±ì„ í•˜ë‚˜ë¡œ
    css = re.sub(r'\s+', ' ', css)
    # { } : ; ì•ë’¤ ê³µë°± ì œê±°
    css = re.sub(r'\s*{\s*', '{', css)
    css = re.sub(r'\s*}\s*', '}', css)
    css = re.sub(r'\s*:\s*', ':', css)
    css = re.sub(r'\s*;\s*', ';', css)
    css = re.sub(r'\s*,\s*', ',', css)
    return css.strip()

def minify_html(html):
    """
    HTMLì„ ì™„ì „íˆ í•œ ì¤„ë¡œ ì••ì¶• (CSS í¬í•¨)
    - <pre>, <textarea> íƒœê·¸ ë‚´ë¶€ë§Œ ë³´ì¡´
    - CSS, Scriptë„ ëª¨ë‘ í•œ ì¤„ë¡œ ì••ì¶•
    """
    # 1. <style> íƒœê·¸ ë‚´ë¶€ CSSë¥¼ ì••ì¶•
    def minify_style_tag(match):
        tag_open = match.group(1)  # <style...>
        css_content = match.group(2)
        minified_css = minify_css(css_content)
        return f"{tag_open}{minified_css}</style>"
    
    html = re.sub(r'(<style[^>]*>)(.*?)</style>', minify_style_tag, html, flags=re.DOTALL | re.IGNORECASE)
    
    # 2. <script> íƒœê·¸ ë‚´ë¶€ë¥¼ í•œ ì¤„ë¡œ ì••ì¶•
    def minify_script_tag(match):
        tag_open = match.group(1)
        script_content = match.group(2)
        # ìŠ¤í¬ë¦½íŠ¸ëŠ” ê°„ë‹¨íˆ ì¤„ë°”ê¿ˆê³¼ ì—°ì† ê³µë°±ë§Œ ì œê±°
        minified = script_content.replace('\n', '').replace('\r', '')
        minified = re.sub(r'\s+', ' ', minified)
        return f"{tag_open}{minified}</script>"
    
    html = re.sub(r'(<script[^>]*>)(.*?)</script>', minify_script_tag, html, flags=re.DOTALL | re.IGNORECASE)
    
    # 3. <pre>, <textarea> íƒœê·¸ ë‚´ë¶€ë¥¼ ì„ì‹œë¡œ ë³´í˜¸
    pre_blocks = []
    def save_pre(match):
        pre_blocks.append(match.group(0))
        return f"___PRE_BLOCK_{len(pre_blocks)-1}___"
    
    html = re.sub(r'<pre[^>]*>.*?</pre>', save_pre, html, flags=re.DOTALL | re.IGNORECASE)
    html = re.sub(r'<textarea[^>]*>.*?</textarea>', save_pre, html, flags=re.DOTALL | re.IGNORECASE)
    
    # 4. HTML ì£¼ì„ ì œê±°
    html = re.sub(r'<!--.*?-->', '', html, flags=re.DOTALL)
    
    # 5. íƒœê·¸ ì‚¬ì´ì˜ ë¶ˆí•„ìš”í•œ ê³µë°± ì œê±°
    html = re.sub(r'>\s+<', '><', html)
    
    # 6. ì—°ì†ëœ ê³µë°±ì„ í•˜ë‚˜ë¡œ
    html = re.sub(r'\s+', ' ', html)
    
    # 7. ì¤„ë°”ê¿ˆ ì œê±°
    html = html.replace('\n', '').replace('\r', '')
    
    # 8. ë³´í˜¸í–ˆë˜ pre/textarea ë¸”ë¡ë“¤ ë³µì›
    for i, block in enumerate(pre_blocks):
        html = html.replace(f"___PRE_BLOCK_{i}___", block)
    
    return html.strip()

def extract_doc_id(filename):
    """íŒŒì¼ëª…ì—ì„œ ë¬¸ì„œID ì¶”ì¶œ (ì˜ˆ: 20151224_ì œëª©_2009465.html -> 2009465)"""
    # .html ì œê±°
    name_without_ext = filename.replace('.html', '')
    # ë§ˆì§€ë§‰ ì–¸ë”ìŠ¤ì½”ì–´ ë’¤ì˜ ìˆ«ì ë¶€ë¶„ì´ ë¬¸ì„œID
    parts = name_without_ext.split('_')
    if len(parts) >= 2:
        # ë§ˆì§€ë§‰ ë¶€ë¶„ì´ ìˆ«ìì¸ì§€ í™•ì¸
        last_part = parts[-1]
        if last_part.isdigit():
            return last_part
    return None

def read_css_files(css_folder):
    """CSS íŒŒì¼ 3ê°œë¥¼ ì½ì–´ì„œ ë”•ì…”ë„ˆë¦¬ë¡œ ë°˜í™˜"""
    css_dict = {}
    css_files = [
        'apr.doc.print.jstl.css',
        'font-load.css',
        'jquery.qtip.min.css'
    ]
    
    for css_file in css_files:
        css_path = css_folder / css_file
        if css_path.exists():
            with open(css_path, 'r', encoding='utf-8', errors='ignore') as f:
                css_dict[css_file] = f.read()
            print(f"âœ“ CSS íŒŒì¼ ì½ê¸° ì™„ë£Œ: {css_file}")
        else:
            print(f"âš ï¸  CSS íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {css_file}")
    
    return css_dict

def insert_css_inline(html_content, css_dict):
    """HTMLì˜ <link> íƒœê·¸ë¥¼ <style> íƒœê·¸ë¡œ êµì²´"""
    soup = BeautifulSoup(html_content, 'html.parser')
    
    # ëª¨ë“  <link> íƒœê·¸ ì°¾ê¸°
    link_tags = soup.find_all('link', rel='stylesheet')
    
    for link_tag in link_tags:
        href = link_tag.get('href', '')
        
        # CSS íŒŒì¼ëª… ì¶”ì¶œ
        css_filename = None
        if 'apr.doc.print.jstl.css' in href:
            css_filename = 'apr.doc.print.jstl.css'
        elif 'font-load.css' in href:
            css_filename = 'font-load.css'
        elif 'jquery.qtip.min.css' in href:
            css_filename = 'jquery.qtip.min.css'
        
        # í•´ë‹¹ CSSê°€ ìˆìœ¼ë©´ <style> íƒœê·¸ë¡œ êµì²´
        if css_filename and css_filename in css_dict:
            style_tag = soup.new_tag('style')
            style_tag.string = css_dict[css_filename]
            link_tag.replace_with(style_tag)
    
    return str(soup)

def main():
    # 1. ê²½ë¡œ ì„¤ì • (ì‚¬ìš©ì í™˜ê²½ì— ë§ê²Œ ìˆ˜ì • í•„ìš”)
    
    base_path = Path("C:/Users/LEEJUHWAN/Downloads/2021-01-01~2025-10-31/html")
    approval_folder = base_path / "ê²°ì¬"
    css_folder = base_path / "resource" / "css"
    
    print(f"ğŸ“‚ ê¸°ë³¸ ê²½ë¡œ: {base_path}")
    print(f"ğŸ“‚ ê²°ì¬ í´ë”: {approval_folder}")
    print(f"ğŸ“‚ CSS í´ë”: {css_folder}\n")
    
    # ê²½ë¡œ ì¡´ì¬ í™•ì¸
    if not approval_folder.exists():
        print(f"âŒ ê²°ì¬ í´ë”ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {approval_folder}")
        print(f"ğŸ’¡ base_path ë³€ìˆ˜ë¥¼ ì‹¤ì œ ê²½ë¡œë¡œ ìˆ˜ì •í•´ì£¼ì„¸ìš”!")
        return
    
    if not css_folder.exists():
        print(f"âŒ CSS í´ë”ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {css_folder}")
        return
    
    # 2. CSS íŒŒì¼ ì½ê¸°
    print(f"ğŸ”„ CSS íŒŒì¼ ì½ëŠ” ì¤‘...\n")
    css_dict = read_css_files(css_folder)
    
    if not css_dict:
        print(f"âŒ CSS íŒŒì¼ì„ ì½ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        return
    
    print(f"\nâœ… {len(css_dict)}ê°œ CSS íŒŒì¼ ë¡œë“œ ì™„ë£Œ\n")
    
    # 3. ì—°ë„ë³„ í´ë”ì—ì„œ HTML íŒŒì¼ ìˆ˜ì§‘
    print(f"ğŸ” HTML íŒŒì¼ ìˆ˜ì§‘ ì¤‘...\n")
    html_files = []
    
    for year_folder in sorted(approval_folder.iterdir()):
        if year_folder.is_dir():
            year = year_folder.name
            year_html_files = list(year_folder.glob("*.html"))
            html_files.extend(year_html_files)
            print(f"   ğŸ“ {year}ë…„: {len(year_html_files)}ê°œ íŒŒì¼")
    
    print(f"\nâœ… ì´ {len(html_files)}ê°œ HTML íŒŒì¼ ë°œê²¬\n")
    
    # 4. HTML íŒŒì¼ ì²˜ë¦¬
    print(f"ğŸ”„ HTML íŒŒì¼ ì²˜ë¦¬ ì‹œì‘...\n")
    result_dict = {}
    success_count = 0
    fail_count = 0
    total_original_size = 0
    total_minified_size = 0
    
    for i, html_file in enumerate(html_files, 1):
        try:
            # ë¬¸ì„œID ì¶”ì¶œ
            doc_id = extract_doc_id(html_file.name)
            if not doc_id:
                print(f"âš ï¸  [{i}/{len(html_files)}] ë¬¸ì„œID ì¶”ì¶œ ì‹¤íŒ¨: {html_file.name}")
                fail_count += 1
                continue
            
            # HTML íŒŒì¼ ì½ê¸°
            with open(html_file, 'r', encoding='utf-8', errors='ignore') as f:
                html_content = f.read()
            
            # CSS inline ì‚½ì…
            html_with_css = insert_css_inline(html_content, css_dict)
            
            # HTML ì••ì¶•
            minified_html = minify_html(html_with_css)
            
            # ë”•ì…”ë„ˆë¦¬ì— ì¶”ê°€
            result_dict[doc_id] = minified_html
            
            success_count += 1
            
            # ì••ì¶• ë¹„ìœ¨ ê³„ì‚°
            original_size = len(html_with_css)
            minified_size = len(minified_html)
            total_original_size += original_size
            total_minified_size += minified_size
            ratio = (1 - minified_size/original_size) * 100 if original_size > 0 else 0
            
            if i % 100 == 0 or i == len(html_files):
                print(f"âœ“ [{i}/{len(html_files)}] ì§„í–‰ ì¤‘... (ì„±ê³µ: {success_count}, ì‹¤íŒ¨: {fail_count})")
        
        except Exception as e:
            print(f"âœ— [{i}/{len(html_files)}] ì—ëŸ¬ ë°œìƒ: {html_file.name}")
            print(f"   ì—ëŸ¬ ë‚´ìš©: {str(e)}")
            fail_count += 1
            continue
    
    # 5. JSON íŒŒì¼ë¡œ ì €ì¥
    output_file = Path("ì „ìê²°ì¬_ë¬¸ì„œIDë³„_HTML_with_CSS.json")
    output_file.parent.mkdir(parents=True, exist_ok=True)
    
    print(f"\nğŸ’¾ JSON íŒŒì¼ ì €ì¥ ì¤‘...")
    with open(output_file, "w", encoding="utf-8") as f:
        json.dump(result_dict, f, ensure_ascii=False, indent=2)
    
    # 6. ê²°ê³¼ ìš”ì•½
    print(f"\n" + "="*60)
    print(f"âœ… ì™„ë£Œ!")
    print(f"="*60)
    print(f"ğŸ“Š í†µê³„:")
    print(f"   - ì „ì²´ HTML íŒŒì¼: {len(html_files)}ê°œ")
    print(f"   - ì„±ê³µ: {success_count}ê°œ")
    print(f"   - ì‹¤íŒ¨: {fail_count}ê°œ")
    print(f"   - ì´ ì›ë³¸ í¬ê¸°: {total_original_size:,} bytes")
    print(f"   - ì´ ì••ì¶• í¬ê¸°: {total_minified_size:,} bytes")
    if total_original_size > 0:
        total_ratio = (1 - total_minified_size/total_original_size) * 100
        print(f"   - ì „ì²´ ì••ì¶•ë¥ : â–¼{total_ratio:.1f}%")
    print(f"\nğŸ“ ì €ì¥ ìœ„ì¹˜: {output_file}")
    print(f"\nğŸ’¡ ì‚¬ìš© ì˜ˆì‹œ:")
    print(f"   import json")
    print(f"   with open('{output_file.name}', 'r', encoding='utf-8') as f:")
    print(f"       data = json.load(f)") 
    print(f"   html = data['2009465']  # ë¬¸ì„œIDë¡œ ë°”ë¡œ ì ‘ê·¼")

if __name__ == "__main__":
    main()


```

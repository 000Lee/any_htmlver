```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
HTML íŒŒì¼ ì¼ê´„ ë³€í™˜ ìŠ¤í¬ë¦½íŠ¸ 

ğŸ“Œ í•µì‹¬ ê¸°ëŠ¥
ì „ìê²°ì¬ HTML íŒŒì¼ì—ì„œ ë‚ ì§œ/ë³´ì¡´ì—°í•œ í˜•ì‹ì„ ì¼ê´„ ë³€í™˜í•˜ëŠ” ìŠ¤í¬ë¦½íŠ¸
ex)
ê¸°ì•ˆì¼: 20210105123307 â†’ 2021-01-05 12:33:07
ë³´ì¡´ì—°í•œ: 3ë…„ 0ì›” â†’ 3ë…„

ğŸ”§ ë™ì‘ ë°©ì‹
ì§€ì •ëœ í´ë”ì—ì„œ .html íŒŒì¼ë“¤ì„ ì¬ê·€ì ìœ¼ë¡œ ì°¾ìŒ
ê° HTMLì—ì„œ <div class="bg01"> íƒœê·¸ë¥¼ íŒŒì‹±
2ë²ˆì§¸ bg01 â†’ ê¸°ì•ˆì¼ ë³€í™˜
4ë²ˆì§¸ bg01 â†’ ë³´ì¡´ì—°í•œ ë³€í™˜
ë³€í™˜ëœ íŒŒì¼ ì €ì¥

âš™ï¸ ì„¤ì • ì˜µì…˜
ROOT_DIRECTORY: ì²˜ë¦¬í•  í´ë” ê²½ë¡œ
TEST_MODE: Trueë©´ 10ê°œë§Œ, Falseë©´ ì „ì²´
SAVE_TO_NEW_FOLDER: Trueë©´ ìƒˆ í´ë”(ì›ë³¸ ë³´ì¡´), Falseë©´ ì›ë³¸ ë®ì–´ì“°ê¸°

ğŸ“¦ ì˜ì¡´ì„±
í‘œì¤€ ë¼ì´ë¸ŒëŸ¬ë¦¬ë§Œ ì‚¬ìš© (os, re, pathlib, datetime, logging) â†’ ë³„ë„ ì„¤ì¹˜ ì—†ìŒ!
"""

import os
import re
import shutil
from pathlib import Path
from datetime import datetime
import logging

# ============================================================
# âš™ï¸ ì„¤ì • (ì—¬ê¸°ë§Œ ìˆ˜ì •í•˜ì„¸ìš”!)
# ============================================================

# ğŸ“ ì²˜ë¦¬í•  ë””ë ‰í† ë¦¬ ê²½ë¡œ
ROOT_DIRECTORY = r"C:\Users\LEEJUHWAN\Downloads\2021-01-01~2025-10-31\html\ê²°ì¬"

# ğŸ§ª í…ŒìŠ¤íŠ¸ ëª¨ë“œ (10ê°œë§Œ ì²˜ë¦¬)
# TEST_MODE = True  # Falseë¡œ ë°”ê¾¸ë©´ ì „ì²´ íŒŒì¼ ì²˜ë¦¬
TEST_MODE = False  # ì£¼ì„ í’€ë©´ ì „ì²´ íŒŒì¼ ì²˜ë¦¬

# ğŸ’¾ ì €ì¥ ë°©ì‹ ì„ íƒ
SAVE_TO_NEW_FOLDER = False  # True: ìƒˆ í´ë”ì— ì €ì¥ (ì›ë³¸ ë³´ì¡´)
                            # False: ì›ë³¸ íŒŒì¼ ì§ì ‘ ìˆ˜ì •

# ğŸ“‚ ìƒˆ í´ë” ì´ë¦„ (SAVE_TO_NEW_FOLDER = Trueì¼ ë•Œë§Œ ì‚¬ìš©)
OUTPUT_FOLDER_NAME = "ê²°ì¬_ë³€í™˜ì™„ë£Œ"

# ============================================================

# ë¡œê¹… ì„¤ì •
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(message)s',
    handlers=[
        logging.FileHandler('html_conversion.log', encoding='utf-8'),
        logging.StreamHandler()
    ]
)

def convert_date_format(date_str):
    """ë‚ ì§œ í˜•ì‹ ë³€í™˜: 20210104151446 -> 2021-01-04 15:14:46"""
    try:
        if len(date_str) != 14 or not date_str.isdigit():
            return None
        year = date_str[0:4]
        month = date_str[4:6]
        day = date_str[6:8]
        hour = date_str[8:10]
        minute = date_str[10:12]
        second = date_str[12:14]
        datetime(int(year), int(month), int(day), int(hour), int(minute), int(second))
        return f"{year}-{month}-{day} {hour}:{minute}:{second}"
    except (ValueError, IndexError):
        return None

def convert_retention_period(period_str):
    """ë³´ì¡´ì—°í•œ í˜•ì‹ ë³€í™˜: 3ë…„ 0ì›” -> 3ë…„"""
    match = re.search(r'(\d+)ë…„\s*\d+ì›”', period_str.strip())
    if match:
        return f"{match.group(1)}ë…„"
    return None

def process_html_content(content):
    """HTML ë‚´ìš© ë³€í™˜"""
    original_content = content
    changes = {'date': None, 'period': None}
    
    # bg01 í´ë˜ìŠ¤ì˜ ëª¨ë“  div ì°¾ê¸°
    pattern = r'<div class="bg01">\s*(.*?)\s*</div>'
    matches = list(re.finditer(pattern, content, re.DOTALL))
    
    if len(matches) < 4:
        return None, changes
    
    # ë‘ ë²ˆì§¸ bg01 (ê¸°ì•ˆì¼) ì²˜ë¦¬
    second_match = matches[1]
    date_text = second_match.group(1).strip()
    converted_date = convert_date_format(date_text)
    
    if converted_date:
        old_div = second_match.group(0)
        new_div = f'<div class="bg01">{converted_date}</div>'
        content = content.replace(old_div, new_div, 1)
        changes['date'] = (date_text, converted_date)
    
    # ë„¤ ë²ˆì§¸ bg01 (ë³´ì¡´ì—°í•œ) ì²˜ë¦¬
    matches_updated = list(re.finditer(pattern, content, re.DOTALL))
    if len(matches_updated) >= 4:
        fourth_match = matches_updated[3]
        period_text = fourth_match.group(1).strip()
        converted_period = convert_retention_period(period_text)
        
        if converted_period:
            new_div = f'<div class="bg01">\n\t\t\t\t\t\t\t{converted_period}\n\t\t\t\t\t\t</div>'
            content = content[:fourth_match.start()] + new_div + content[fourth_match.end():]
            changes['period'] = (period_text, converted_period)
    
    if content == original_content:
        return None, changes
    
    return content, changes

def process_file(file_path, output_path):
    """ë‹¨ì¼ íŒŒì¼ ì²˜ë¦¬"""
    try:
        # íŒŒì¼ ì½ê¸°
        with open(file_path, 'r', encoding='utf-8') as f:
            content = f.read()
        
        # ë‚´ìš© ë³€í™˜
        new_content, changes = process_html_content(content)
        
        if new_content is None:
            return False, changes
        
        # íŒŒì¼ ì €ì¥
        output_path.parent.mkdir(parents=True, exist_ok=True)
        with open(output_path, 'w', encoding='utf-8') as f:
            f.write(new_content)
        
        return True, changes
        
    except Exception as e:
        logging.error(f"âŒ ì˜¤ë¥˜: {file_path.name} - {str(e)}")
        return False, {}

def main():
    root_path = Path(ROOT_DIRECTORY)
    
    if not root_path.exists():
        print(f"âŒ ë””ë ‰í† ë¦¬ê°€ ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤: {ROOT_DIRECTORY}")
        return
    
    # ì¶œë ¥ ë””ë ‰í† ë¦¬ ì„¤ì •
    if SAVE_TO_NEW_FOLDER:
        output_root = root_path.parent / OUTPUT_FOLDER_NAME
        output_root.mkdir(exist_ok=True)
        mode_text = f"ìƒˆ í´ë”ì— ì €ì¥: {output_root}"
    else:
        output_root = root_path
        mode_text = "ì›ë³¸ íŒŒì¼ ì§ì ‘ ìˆ˜ì •"
    
    # HTML íŒŒì¼ ì°¾ê¸°
    html_files = list(root_path.rglob('*.html'))
    total_files = len(html_files)
    
    if total_files == 0:
        print("âš ï¸  ì²˜ë¦¬í•  HTML íŒŒì¼ì´ ì—†ìŠµë‹ˆë‹¤.")
        return
    
    # í…ŒìŠ¤íŠ¸ ëª¨ë“œ ì ìš©
    if TEST_MODE:
        html_files = html_files[:10]
        process_count = len(html_files)
        test_text = f"ğŸ§ª í…ŒìŠ¤íŠ¸ ëª¨ë“œ: {process_count}ê°œ íŒŒì¼ë§Œ ì²˜ë¦¬"
    else:
        process_count = total_files
        test_text = f"âœï¸  ì „ì²´ ëª¨ë“œ: {process_count}ê°œ íŒŒì¼ ì²˜ë¦¬"
    
    # ì‹œì‘ ë©”ì‹œì§€
    print("\n" + "="*70)
    print("HTML íŒŒì¼ ë³€í™˜ ìŠ¤í¬ë¦½íŠ¸")
    print("="*70)
    print(f"ğŸ“ ì›ë³¸ ê²½ë¡œ: {root_path}")
    print(f"ğŸ’¾ ì €ì¥ ë°©ì‹: {mode_text}")
    print(f"ğŸ“Š {test_text}")
    print("="*70)
    
    response = input("\në³€í™˜ì„ ì‹œì‘í•˜ì‹œê² ìŠµë‹ˆê¹Œ? (y/n): ")
    if response.lower() != 'y':
        print("ì‘ì—…ì´ ì·¨ì†Œë˜ì—ˆìŠµë‹ˆë‹¤.")
        return
    
    print("\nğŸš€ ë³€í™˜ ì‘ì—… ì‹œì‘...\n")
    
    success_count = 0
    fail_count = 0
    
    for idx, file_path in enumerate(html_files, 1):
        # ì¶œë ¥ ê²½ë¡œ ê²°ì •
        if SAVE_TO_NEW_FOLDER:
            relative_path = file_path.relative_to(root_path)
            output_path = output_root / relative_path
        else:
            output_path = file_path
        
        # íŒŒì¼ ì²˜ë¦¬
        success, changes = process_file(file_path, output_path)
        
        if success:
            success_count += 1
            print(f"âœ… [{idx}/{process_count}] {file_path.name}")
            if changes.get('date'):
                old, new = changes['date']
                print(f"   ğŸ“… ê¸°ì•ˆì¼: {old} â†’ {new}")
            if changes.get('period'):
                old, new = changes['period']
                print(f"   ğŸ“‹ ë³´ì¡´ì—°í•œ: {old} â†’ {new}")
        else:
            fail_count += 1
            if fail_count <= 3:  # ì²˜ìŒ 3ê°œë§Œ í‘œì‹œ
                print(f"âš ï¸  [{idx}/{process_count}] {file_path.name} - ë³€ê²½ì‚¬í•­ ì—†ìŒ")
    
    # ê²°ê³¼ ìš”ì•½
    print("\n" + "="*70)
    print("ğŸ“Š ë³€í™˜ ì™„ë£Œ!")
    print("="*70)
    print(f"ì²˜ë¦¬ëœ íŒŒì¼: {process_count}ê°œ")
    print(f"âœ… ì„±ê³µ: {success_count}ê°œ")
    print(f"âš ï¸  ì‹¤íŒ¨/ë³€ê²½ì—†ìŒ: {fail_count}ê°œ")
    if SAVE_TO_NEW_FOLDER:
        print(f"\nğŸ’¾ ë³€í™˜ëœ íŒŒì¼ ìœ„ì¹˜: {output_root}")
        print(f"ğŸ“ ì›ë³¸ íŒŒì¼ì€ ê·¸ëŒ€ë¡œ ë³´ì¡´ë©ë‹ˆë‹¤: {root_path}")
    if TEST_MODE:
        print(f"\nğŸ’¡ í…ŒìŠ¤íŠ¸ ëª¨ë“œì˜€ìŠµë‹ˆë‹¤. TEST_MODE = Falseë¡œ ë°”ê¾¸ë©´ ì „ì²´ íŒŒì¼ì„ ì²˜ë¦¬í•©ë‹ˆë‹¤.")
    print("="*70)

if __name__ == "__main__":
    main()
```

```python
# 기틀 json의 docBody에 통 html 삽입 

import json

# 파일 읽기
with open('결재목록_변환결과.json', 'r', encoding='utf-8') as f:
    approval_list = json.load(f)

with open('전자결재_문서ID별_HTML.json', 'r', encoding='utf-8') as f:
    html_dict = json.load(f)

# docBody 교체
for item in approval_list:
    source_id = item.get('sourceId', '')
    if source_id.startswith('Prefix_'):
        doc_id = source_id.replace('Prefix_', '')
        if doc_id in html_dict:
            item['docBody'] = html_dict[doc_id]

# 결과 저장
with open('결재목록_교체완료.json', 'w', encoding='utf-8') as f:
    json.dump(approval_list, f, ensure_ascii=False, indent=2)
```

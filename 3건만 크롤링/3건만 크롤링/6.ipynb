```python
# 최종 cmds로 만들기 위한 txt maker
# json을 addDocument{}한줄로 만들어줍니다.
# 노트플러스에서 ' 빼고, 확장자 cmds로 바꿔주기 

import json

# JSON 파일 읽기 - 파일 경로를 실제 파일명에 맞게 수정하세요
input_file = '결재목록_업데이트완료.json'  # 또는 전체 경로 입력
output_file = 'output_commands.txt'

try:
    with open(input_file, 'r', encoding='utf-8') as f:
        data = json.load(f)
    
    # 출력 파일 생성
    with open(output_file, 'w', encoding='utf-8') as out:
        for item in data:
            # JSON 객체를 문자열로 변환 (한글 유지, 들여쓰기 없음)
            json_str = json.dumps(item, ensure_ascii=False, separators=(',', ':'))
            
            # addDocument 명령어 형식으로 작성
            command = f"addDocument {json_str}\n"
            out.write(command)
    
    print(f"✓ 성공: 총 {len(data)}개의 addDocument 명령어가 생성되었습니다.")
    print(f"✓ 출력 파일: {output_file}")
    
except FileNotFoundError:
    print(f"✗ 오류: '{input_file}' 파일을 찾을 수 없습니다.")
    print("현재 작업 디렉토리에 JSON 파일이 있는지 확인하거나,")
    print("코드의 input_file 변수에 전체 경로를 입력하세요.")
    print("\n예시:")
    print("  input_file = r'C:\\Users\\사용자명\\Downloads\\결재목록_변환결과 (3).json'")
except json.JSONDecodeError:
    print(f"✗ 오류: '{input_file}' 파일이 올바른 JSON 형식이 아닙니다.")
except Exception as e:
    print(f"✗ 오류 발생: {e}")
```
